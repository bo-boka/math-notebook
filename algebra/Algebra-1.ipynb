## Exponentiation
* Signifies repeated multiplication of a number upon itself. The number to be repeatedly multiplied is called the **base** and the amount of times it's to be multiplied with itself is called the **exponent** or the **power**. 
* Written either by separating the base and exponent with a carat, b^n, or with the power as a superscript, $b^n$ where b is the base and n is the exponent/power. Pronounced as "b to the power of n".
* **Perfect Square** is any whole number taken to the power of 2.

#### Simplifying Exponents
* Generally, if the exponent $\leq 3$, apply the exponent. If the exponent $\geq3$, leave the answer in exponential form.
* Any number with an exponent of 0 results in 1.
 * Example: $5^0 = 1$
* For negative numbers with exponents that aren't wrapped in parenthesis, the sign of the result will be negative.
 * Example: $-3^4 = -(3 * 3 * 3 * 3) = -81$
* For negative numbers with exponenets wrapped in parenthesis, the sign of the result is positive if the exponent is even and negative if the exponent is odd.
 * Example: $(-3)^4 = (-3)(-3)(-3)(-3) = 81$
 * Example: $(-3)^3 = (-3)(-3)(-3) = -27$
* **Product Property of Exponents**: When multiplying exponents where the bases are the same, add the exponents. I.e. If a is a real number and m,n are counting numbers, then $a^m * a^n = a^{m+n}$
 * Example: $2^2 2^3 = 2^{2+3} = 2^5$
* **Power Property of Exponents**: When raising a power to a power, multiply the exponents. I.e If a is a real number and m,n are whole numbers, then $(a^m)^n = a^{mn}$
 * Example: $(3^4)^2 = 3^{4 * 2} = 3^8$
* **Product to a Power Property**: When multiplying exponents where the bases are different, apply the exponent to both bases. I.e. If a and b are real numbers and m is a whole number, then $(ab)^m = a^m b^m$
 * Example: $(2 * 3)^2 = 2^2 3^2 = 36$
* **Quotient Property of Exponents**: When dividing exponents where the bases are the same, if the denominator exponent is smaller, subtract it from the numerator. If the numerator exponent is smaller, subtract it from the denominator & put the result over 1. I.e. If a is a real number, a$\ne$0, and m,n are whole numbers, then if m > n $\frac{a^m}{a^n} = a^{m-n}$, else if m < n $\frac{1}{a^{n-m}}$
 * Example: $\frac{2^5}{2^3} = 2^{5-3} = 2^2$
 * Example: $\frac{2^3}{2^5} = \frac{1}{2^{5-3}} = \frac{1}{2^2}$
* **Negative Exponents**: Represents the case of Quotient Property of Exponents where the numerator exponent is smaller than the denominator. Instead of putting the altered result over 1, we use a negative exponent, although that's not its simplest form. I.e. If n is a positive integer and a$\ne$0, then $\frac{1}{a^n} = a^{-n}$
 * Example: $4^{-2} = \frac{1}{4^2}$
* **Quotient to a Power Property**: For wrapped fractions with an exponent, apply the exponent to both the numerator & denominator. I.e. If a and b are real numbers, b$\ne$0, and m is a counting number, then $(\frac{a}{b})^m = \frac{a^m}{b^m}$
 * Example: $(\frac{7}{8})^2 = \frac{7^2}{8^2} = \frac{49}{64}$
 
#### Scientific Notation
* A convenient, shorthand way of writing very large or very small numbers to make calculation easier, often in science.
* A number is written in scientific notation when a number between 0 and 10 (1-9) is multiplied by a power of 10. 
 * Customary to use the multiplication sign, despite that it's not generally used in algebra.
 * If the original number is:
   * \> 1, the power will be $10^n$ (positive)
   * < 1, the power will be $10^{-n}$ (negative)
* To **convert large numbers**, move the decimal to the left until it's a number between 0-10, exclusive, & multiply that number by 10 to the power of the amount of times the decimal was moved.
 * Example: $650,000,000 = 6.5 * 10^8$
* To **convert very small decimal numbers**, move the decimal to the right until it's a number between 0-10, exclusive, & multiply that number by 10 to the negative power of the amount of times the decimal was moved. 
 * Small # Example: $0.004 = 4 * 10^{-3}$
* To **multiply & divide expressions**, use the Properties of Exponents.
 * Example multiplication: $(4 * 10^5)(2 * 10^{-7}) = (4 * 2)(10^{5+(-7)}) = 8 * 10^{-2} = 0.08$
 * Example division: $\frac{9 * 10^3}{3 * 10^{-2}} = (\frac{9}{3})(10^{3-(-2)}) = 3 * 10^5 = 300,000$
 
## Roots
* The reversal of exponentiation. Represents a number called the **root** that when multiplied by itself a given number of times, called the **degree** or **index**, will equal another given number, called the **radicand**. 
* Written using a radical symbol, $\sqrt[n]{x}$, where n is the degree and x is the radicand. Pronounced as "nth root of x".
* **Square Root** is a number multiplied by itself only once to get a square are used most commonly such that the radical symbol without a specified degree is implied to be a square root operation, $\sqrt{x}$. Pronounced as "square root of x" rather than "2nd root of x". See below for more detail.
* **Cube Root** is a number multiplied by itself twice. Written as $\sqrt[3]{x}$ and pronounced as "cube root of x" rather than "third root of x".

#### Square Roots
* aka 'positive square root' or 'principal square root'. 
* Every positive number has two square roots, a positive one and a negative one. 
 * The 'Positive Square Root', aka the 'Pricipal Square Root' is represented with the radical symbol, $\sqrt{x}$. 
 * The negative square root is represented by placing a minus symbol in front of  the radical sign, $-\sqrt{x}$
* Cannot take a square root of a negative number because two positives multiplies are positive while two negatives multiplied also equal a positive. Rather, there is no "Real Number" solution (See Types of Numbers below).
* For square roots that aren't perfect squares, we can either use relationship symbols to represent an approximation of value, or, if we use a calculator, the answer will be an approximate decimal and will require the approximation symbol (rather than the equals sign) and usually round to the nearest hundrenth decimal.
 * Example: $2 < \sqrt{5} < 3$ 
 * Example: $\sqrt{5} \approx 2.24$
* Area & Side Length of Squares: Since the area of a square is calculated by squaring the length of any of its sides, likewise the length of a square's side(s) can be calculated by taking the square root of its area, $\sqrt{A}$
 * Example: If the Area of a square = 9, then the length of its sides is $\sqrt{9} = 3$
* Gravity: On Earth, if an object is dropped from a height of h feet, the time in seconds it will take to reach the ground is found by evaluating the expression $\frac{\sqrt{h}}{4}$
 * Example: If an object is dropped from a height of 64 feet, the amount of seconds for it to reach the ground is $\frac{\sqrt{64}}{4} = \frac{8}{4} = 2$ seconds.
* Car Accident Investigation: Using square root formula with skid mark distance can determine the speed the car was going before applying the brakes. According to some formulas, if the length of the skid marks is d feet, then the speed of the car, in mph, can be found by evaluating $\sqrt{24d}$
 * Example: If skid marks were 190 ft, the speed of the car (in mph) before braking is $\sqrt{24(190)} = \sqrt{4560} = 67.527772...$. The rounded speed is 67.5 mph


# Types of Numbers

## Main Sets
* **Natural Numbers** (N): (aka counting numbers or positive integers) are the numbers {1, 2, 3, 4, 5, …}
* **Whole Numbers** (W): is the set of  natural numbers, plus zero, i.e. {0, 1, 2, 3, 4, 5, …}.
* **Integers** (Z): set of all whole numbers plus all the negatives (or opposites) of the natural numbers, i.e. {… , ⁻2, ⁻1, 0, 1, 2, …}
* **Rational Numbers** (Q): any number that can be represented as a fraction, where the top and bottom numbers are integers & the denominator is not 0, e.g. 1/2, 3/4, 7/2, ⁻4/3, 4/1 
 * This includes all integers since they can be represented as a fraction with a denominator of 1 in the case of non-zero numbers, and a fraction with the numerator of 0 in the case of 0.
 * This also includes decimals that end or have repeating digits.
 * But does not include decimals with infinite non-repeating digits.
 * Notice that "ratio" is in the name.
* **Irrational Numbers**: numbers that cannot be expressed as a fraction, e.g. $\pi$, $\sqrt2$, $\sqrt{3}$, n
* **Real Numbers** (R): (also called measuring numbers or measurement numbers). Includes all rational, & irrational numbers. Every real number corresponds to a point on the number line.
* **Complex Numbers** (C): numbers which have both real and imaginary part.

#### Discrete vs Continuous Numbers
The above types of number sets can be split up into discrete or continuous numbers.
* N, W, Z and Q are referred to as discrete. This means that they are separate and distinct entities & each of these sets is countable.
* The last set, R, cannot be counted. This is because they are continuous. Between any two real numbers, however close they may be, there are infinitely more real numbers.

# Integers
* Type of rational number, comprising of all whole numbers (which include natural numbers) as well as the negatives of the natural numbers.

## Types of Integers
* **Even Numbers**: Numbers that are multiples of 2.
* **Odd Numbers** numbers: Numbers which are not multiples of 2 & leave 1 as the remainder when divided by 2.
* **Prime Numbers**: Any number other than 1 which does not have any factor apart from one and the number itself is called a prime number.
 * **Co-prime/Relatively Prime Numbers**: Two numbers are said to be co-prime or relatively prime if they do not have any common factor other than one.
* **Composite Numbers**: a number that has more than two distinct factors.
* **Perfect Numbers** are equal to the sum of their “proper” divisors (positive integers that divide a number evenly, not counting itself). For example, 6 = 3 + 2 + 1, and 28 = 14 + 7 + 4 + 2 + 1


## Concept: Multiples & Factors
* A **factor** of a number is the number that divides it completely without leaving any remainder.
 * Two numbers that we multiply together to get a certain product are called **factor pairs**. 
* A **multiple** of a number is the product of the number and a counting number
* Notice that the Types of Integers are relative to their factors & multiples.

#### Pattern of Multiples
* 2: even numbers
* 3: has a sum of digits that is multiple of 3
* 5: last digit is either 0 or 5
* 6: is divisible by both 2 and 3
* 10: the last digit is 0

#### Factorization
* Determining all the factors of a number
* We can find all the factors of any counting number by systematically dividing the number by each counting number, starting with 1. If the quotient is also a counting number (whole number), then the divisor and the quotient are factors of the number. We can stop when the quotient becomes smaller than the divisor because the factors will start to duplicate.

#### Determining Primality
* Determining if a number is a prime number.
* Whole Numbers are either prime or composite, with the exception of 1, which is neither. All composite numbers have prime factors. So, if a number has a prime factor, it's not a prime & is instead a composite.
* For quick determination of some numbers, we can use patterns of multiples for 2, 3, & 5. 
 * Even numbers (except 2) are composite.
 * If the sum of the number's digits is a multiple of 3, it's composite.
 * If the number ends in 5 or 0, it's composite.
* For complete determination, we use the same process used in factorization except divide the number by prime numbers until the quotient becomes smaller than the divisor. If no primes divide evenly, the number is prime.

#### Prime Factorization
* every composite number can be written as a unique product of primes
* can either be written by specifying each prime separated by a mulitplication sign or in exponential form.
* find prime factorization of a number using the ladder method
 * To begin building the “ladder,” divide the given number by its smallest prime factor. To add a “step” to the ladder, we continue dividing by the same prime until it no longer divides evenly. Then we divide by the next prime. Continue dividing up the ladder in this way until the quotient is prime.
 * Example: Find Prime Factorization of 36
   * 36/2=18, 18/2=9, 9/3= 3
   * Take all the divisors & the last quotient: 2 * 2 * 3 * 3 = $2^2 * 3^2$
   
#### Least Common Multiple (LCM)
* A common multiple of two numbers is a number that is a multiple of both numbers. The LCM is the smallest of the common multiples of those numbers.
* Useful when adding/subtracting fractions with different denominators.
* Find the LCM in smaller numbers by seeing if the smaller number is a divisor of the larger. If so, the larger number is the LCM. If not, multiply the larger number by 2 and check again.
* Find the LCM in small or large numbers by using the Prime Factor Method. To do this, find the Prime Factorization of both numbers. If there are repeating factors between numbers, we're only using each factor once. That is, use repeated factors from the same number, but don't take those factors again from the second number (can use columns to determine this). Then multiply those factors together. The product is the LCM.
 * Exmaple: Find the LCM of 12 and 18
   * 12 = 2 * 2 * 3 ....
   * 18 = 2 * .... 3 * 3
   * Since both numbers have both 2's & 3's, we're only taking the largest amount for each. 12 has two 2's whereas 18 only has 1, so we keep 2 * 2. Likewise, 18 has two 3's whereas 12 only has one 3. so we keep 3 * 3. You can also think of these as columns where if you dropped down the numbers from each column, you'd get:
   * 2 * 2 * 3 * 3 = 36

#### Greatest Common Factor (GCF)
* Found similarly to finding the LCM, except after the columns are lined up, only keep the factors that occur in both sets.
 * Example: Use above for 12 and 18
   * 2 and 3 are the duplicates.
   * 2 * 3 = 6

## Operations on Negative Integers
* Integers, unlike whole numbers (which are natural numbers and 0), can be negative. Another name for the set of natural numbers is positive integers.
* Operations on negative integers have a different set of rules that affects their sign.
* If the negative number comes after the operator sign, it's surrounded in parenthesis.
* If the number(s) are surrounded in parenthesis with no operator sign, the operator is multiplication.

#### Addition of Negative Integers
* If the second number is a negative, the operator becomes subtraction
 * Example: 5 + (-2) = 3
* 2 negative sum a negative
 * Example: -5 + (-2) = -7
* When summing a positive & negative, if there is more negatives, sum is negative whereas more positives, the sum is positive.

#### Subtraction of Negative Integers
* If the second number is a negative, the operator becomes addition (2 negatives = a positive)
 * Example: 5 - (-2) = 7
 * Exmaple: -5 - (-2) = -3

#### Division & Multiplication of Negative Integers
* In algebra, multiplication is usually represented using parenthesis.
* Multiplying/Dividing 2 negatives, the product/quotient is positive.
 * Example: -2(-4) = 8
* Multiplying/Dividing 1 negative & 1 positive, the product/quotient is negative.
 * Example: -2(4) = -8
* Can change the sign of a number by multiplying or dividing it by -1.
 * Example: -16(-1) = 16

# Fractions
* Fractions are a type of rational number that represent a part(s) of a whole, and are of the form a/b, where both a are whole numbers, but b$\neq$0. 
* Formatted as $\frac{a}{b}$, where a is the **numerator** representing a number of equal parts, and b is the **denominator**, which represents all the equal parts creating a whole.
* Fractions are sometimes defined as having the numerator less than the denominator & written in their lowest terms.
* Fractions also represent division where the numerator is the dividend, the denominator is the divisor, and the fraction bar is the division symbol. 


## Types of Fractions
* **Proper Fraction**: Numerator is less than the denominator.
* **Improper Fraction**: Numerator is greater than the denominator.
* **Mixed Fraction**: A natural number next to a proper fraction.
* **Unit Fraction**: Numerator is 1 and denominator is positive integer greater than 1.
* **Simplified Fraction**: Lowest terms; There are no common factors, other than 1, between the numerator & denominator.
* **Complex Fractions**: Numerator and/or the denominator contains a fraction.
* **Equivalent Fractions**: Fractions that have the same value despite having different numerators & denominators. When simplified, they are the same number.
* **Like Fractions**: Two fractions with the same denominators.
* **Unlike Fractions**: Two fractions with different denominators.
* **Decimal Fraction**: Denominator has powers of 10.


#### Reducing Fractions
* Simplifying fraction to its lowest terms.
* Essentially uses the Equivalent Fractions Property in reverse (see Equivalent Fractions below).
* Use Prime Factorization to reduce fractions.
 * Do Prime Factorization on the numerator & denominator.
 * Remove the common factors of each
 * Multiply the leftover factors
 * Example: Reduce $\frac{18}{24}$
   * Prime Factorize numerator: 18 = 2 * 3 * 3
   * Prime Facrorize denominator: 24 = 2 * 2 * 2 * 3
   * Remove commons: $\frac{3}{2 * 2}$
   * Multiply leftovers: $\frac{3}{4}$
* When reducing improper fractions, do not change it to a mixed number

#### Least Common Denominator (LCD)
* LCD is essentially finding the LCM, but for the denominators of fractions (see LCM above).
* We use LCD to create Equivalent Fractions to perform operations (see Equivlent Fractions below).

#### Equivalent Fractions
* Can create equivalent fraction to a given fraction by using the Equivalent Fractions Property, which states that "If a,b, and c are numbers where $b\neq0$ and $c\neq0$, then $\frac{a}{b} = \frac{ac}{bc}$
* Multiplying the numerator & denominator by the same number gives an equivalent fraction.

#### Unlike Fractions to Like Fractions
* If we need to perform addition or subtraction operations on Unlike Fractions, we first need to convert them to Like Fractions so that their denominators match. 
* Conversion can be done by first finding the LCD & creating Equivalent Fractions based on the LCD for one or both of the fractions.
 * Example: Convert $\frac{1}{4}$ and $\frac{1}{6}$ to Like Fractions with method 1.
 * Find the LCD: 
   * Factors of 4 = 2 * 2
   * Factors of 6 = 2 * 3
   * Unique factors = 2 * 2 * 3 = 12
 * Multiply each fraction's numerator by the same number used to multily the denominator to get 12: $\frac{1 * 3}{4 * 3} = \frac{3}{12}$ and $\frac{1 * 2}{6 * 2} = \frac{2}{12}$
* Conversion can also be done by using the missing factor after dropping the columns while finding the LCD. Multiply the missing factor by their fraction's numerator & denominator.
 * Example: Convert $\frac{1}{4}$ and $\frac{1}{6}$ to Like Fractions with method 2.
 * Find the LCD:
   * Factors of 4 = 2 * 2 ....
   * Factors of 6 = 2 * .... 3
   * Missing column(....) factor for 4 is 3
   * Missing column(....) factor for 6 is 2
  * Multiply 4th fraction by 3 and the 6th fraction by 2: $\frac{1 * 3}{4 * 3}$ and $\frac{1 * 2}{6 * 2}$
    
#### Improper Fractions & Mixed Numbers
* Improper Fraction to Mixed Numbers: Divide the fraction. The number of times it divides evenly is the natural number. The remainder becomes the numerator in the accompanying fraction.
* Mixed Number to Improper Fraction: Multiply the denominator by the natural number & place the product in the numerator.

#### Reciprocals
* Can invert a fraction's numerator & denominator if neither are 0 to get the reciprocal
 * Example: The reciprocal of $\frac{2}{5}$ is $\frac{5}{2}$
* The reciprocal of a negative is also a negative.
* Multiplying a fraction by its reciprocal gives a product of 1.

#### Decimal Fractions
* If the denominator of a fraction can be written in the form of prime factorization consisting of either 2's, 5's, or both, it means it can be converted to a decimal fraction. 
* To convert a fraction to a deimal fraction, convert it to an equivalent fraction that has a power of 10 in the denominator. 
* Find a power of 10 for the denominator by dividing the denominator by powers of 10 until they divide evenly. Multiply the quotient by the numerator.
 * Example: $\frac{9}{4} = \frac{9 * 25}{4 * 25} = \frac{225}{100}$


## Operations on Fractions
You need a common denominator to add or subtract fractions, but not to multiply or divide.

#### Adding/Subtracting Fractions
* If the fractions are Like Fractions, only add the numerators & keep the same denominator.
 * Example: $\frac{1}{5} + \frac{2}{5} = \frac{3}{5}$
* If the fractions are Unlike Fractions, convert them to like fractions & add the numerators.
* For Mixed Numbers that have Unlike Fractions &/or the operation is subtraction, it's best to convert to improper fractions first.

#### Multiplying Fractions
* Multiply the numerator with the numerator and the denominator with the denominator.
 * Example: $\frac{1}{2} * \frac{2}{5} = \frac{2}{10} = \frac{1}{5}$
* When working with negatives, the same rules apply as with integers.
 * Example: $-\frac{5}{8}(-\frac{2}{3}) = \frac{10}{24} = \frac{5}{12}$
* When multiplying a fraction with an integer, it's helpful to write the integer as a fraction.
 * Example: $\frac{1}{7} * 56 = \frac{1}{7} * \frac{56}{1} = \frac{56}{7} = 8$

#### Dividing Fractions
* Multiply the first fraction by the reciprocal of the second.
 * Example: $\frac{2}{3} \div \frac{5}{6} = \frac{2}{3} * \frac{6}{5} = \frac{12}{15} = \frac{4}{5}$
* When evaluating Complex Fractions, it's helpful to rewrite the fraction as an expression where the main fraction bar is a division symbol & then multiply the reciprocal accordingly.
 * Example: $\frac{\frac{3}{4}}{\frac{5}{8}} = \frac{3}{4} \div \frac{5}{8} = \frac{3}{4} * \frac{8}{5} = \frac{24}{20} = \frac{6}{5}$
* Since a fraction represents division, a negative sign within the fraction is important. Most often, negative symbols will be before the entire fraction but they can appear in the numerator, the denominator, or both. A single negative regardless of placement will render a negative quotient. If negatives are in both the numerator & denominator, the quotient is positive.

# Decimals
* Decimals a type of rational or irrational number, which has a whole number and the fractional part separated by a decimal point, which are represented in base 10. 
* All fractions are decimals but not all decimals are fractions. Fractions are rational numbers whereas decimals can be rational or irrational (e.g. pi is irrational)
 * Decimals are another way of writing fractions whose denominators can be written in powers of ten, Example: $\frac{9}{4} = \frac{225}{100} = 2.25$
* Writing a number as a decimal is known as decimal notation.
 * Note that decimals are usually rounded to the thousandth, unless otherwise specified or when working with money, in which case round to the hundredth.
* Since the right of the decimal represents values less than 1, the place values grow smaller as you move to the right ascending each place in powers of 10. Note the figure below:
 * The “th” at the end of the name means the number is a fraction. “One thousand” is a number larger than one, but “one thousandth” is a number smaller than one.
 * The tenths place is the first place to the right of the decimal, but the tens place is two places to the left of the decimal.


![title](./../resources/dec-places.png)

## Types of Decimals
* **Terminating Decimals**: aka exact Decimal Number. Has a finite, countable number of digits just after the decimal. Ex, 1.25
* **Recurring Decimals**: Non-Terminating decimal where digits after the decimal point repeat at regular intervals & continue on endlessly. Can be written in fractional form & therefore are Rational Numbers. Can be pure periodic or ultimately periodic. Repeating decimals are represented with a straight line above recurring digit pattern. Ex, $3.123123123... = 3.\overline{123}$
* **Non-Recurring Decimals**: Non-Terminating decimal having an infinite number of digits that follow no specific order. Cannot be written as a fraction & are therefore Irrational Numbers. Represented with an ellipsis (...) toward the end. Ex, 3.142857143... 

#### Decimals to Fractions
* A decimal with a zero to the left of the decimal point converts to a proper fraction whereas a decimal with a counting number to the left of the point converts to a mixed number.
* To convert the digits on the right of the decimal point, correlate the place value to the denominator & place the number as the numerator, then simplify.
 * Example: $3.24 = 3\frac{24}{100} = 3\frac{6}{25}$
* To convert recurring decimals that have only repeating digits to right of the decimal point, place the repeating sequence in the numerator & place an amount of 9's in the denominator according to the number of sequence digits.
 * Example: $0.\overline{123} = \frac{123}{999}$
* To convert recurring decimals with a unique digit followed by recurring digits, move the decimal to the right until hitting the recurring digits. Then minus the original number from the new number (the repeating digits cancel out). The put the number of whole digits as 9's in the denominator. To get rid of the decimal, multiply top & bottom by 10 as many times as necessary.
 * Example: $0.1\overline{6} = \frac{1.5}{9} = \frac{15}{90} = \frac{1}{6}$

#### Fractions to Decimals
* The fraction bar is a division symbol.
* Convert by carring out the division between the numerator by the denominator.
 * If the quotient continues for many decimal places, it is often rounded. The number of places rounded will depend on the situation, but usually rounded to the thousandth place (3 decimal places). When dealing with money, round to the hundredths place.

#### Removing Decimals From a Fraction
* When decimals occur in the numerator or denominator, convert them to whole numbers by eliminate the decimals using the Equivalent Fractions Property and multiplying by powers of 10, aka moving the decimals the same amount of times on the top & bottom. Then reduce.
   * Example: $\frac{0.8}{0.05} = \frac{(0.8)100}{(0.05)100} = \frac{80}{5}$

#### Powers of 10
* In math, it's common to multiply decomals by powers of 10.
* When you multiply by a power of 10, the decimal moves to the right as many times as the power has 0's.
 * Example: 45.86 * 100 = 4586
 
## Operations on Decimals
Operations involve manipulating or removing the decimals, leaving long numbers to operate on. Become familiar with Long Addition/Subtraction/Multiplication/Division.

#### Adding/Subtracting Decimals
* When adding/subtracting decimals, add zeros to the end of the place values until they each have the same amount of digits
* When subtracting decimals from a whole number, subtract right digits from the place value & minus an extra 1.
* When subtracting larger decimals from smaller decimals, switch them to subtract the smaller from the larger & add a negative sign to the result.

#### Multiplying Decimals
* For decimals with a zero on the left, count the sum of places in each decimal. Multiply the digits and place them at the end of the place count sum.
 * Example: (0.2)(0.46)
   * Count the right digits & add them together: 3
   * Multiply the right digits: 2(46) = 92
   * Place that at the end of the digit count(3): 0.092
* For decimals with counting numbers on the left, count the sum of decimal places in each. Remove the decimals & multiply as if they were whole numbers. Place the decimal back counting the place sum from the left.
 * Example: (3.9)(4.075)
   * Count the right digit places for both: 4
   * Drop the decimals & multiply regularly: 39 * 4075 = 158925
   * Add the decimal back per the sum of places: 15.8925
   
#### Dividing Decimals
* Use long division format. Move the divisor decimal to make it a whole number. Move the dividend decimal the same number of times. Bring the decimal straight up. Divide as usual.

# Concept: Ratios
* A ratio compares two or more quantities, called **terms**, that are measured with the same unit. Pronounced as "a to b". Written as either $\frac{a}{b}$ or a:b
 * When expressed using fractional notation, the fraction should be simplified. If the ratio is an improper fraction, we don't change it to a mixed number or a whole number because we're illustrating a 2 part comparison.
* All fractions are ratios, but not all ratios are fractions, despite that they can be formatted as fractions.
 * Fractions are ratios that compare a number of units to the total number of units, called a **part-to-whole ratio**
 * But generally, ratios compare a number of units in a whole to another number of units in a whole, called a **part-to-part ratio**
 * Example: There are 7 total marbles, 2 of which are red, and 5 of which are blue.
   * The ratio of red to blue is 2:5
   * The ratio of red to total is 2:7
* Ratios can compare more than two quantities.
 * Ratios that compare 2 quantites of the same group are called **two-term ratio**.
 * Ratios that compare 3 quantities of the same group are called **three-term ratio**.
 * Example: There are 7 total marbles, 2 are red, 4 are blue, 1 is green.
   * Two-term ratio comparing red to green: 2:1
   * Three-term ratio comparing red to blue to green: 2\:4:1 
* **Rate**: type of ratio that compares two quantities that are measured with different units. Pronounced as "a units per b units". Written usually as a fraction.
 * Units need to be specified.
 * When simplifying the fraction, the unit remains in both the numerator & denominator.
 * Usually want to represent rate with a **Unit Rate**, meaning the number 1 is in the denominator, e.g. hourly pay rate, driving rate per hour, unit price, etc.
 * Rates can be converted to Unit Rates by dividing the numerator by the denominator.
   * Examples: $\frac{384 dollars}{32 hours} =$ 12 dollars per hour


## Proportions
* Proportion is the relation or equality between two ratios or fractions.
* Formatted as an equation of fractions, $\frac{a}{b} = \frac{c}{d}$ where $b \neq 0, d \neq 0$, and read as "a is to b, as c is to d".
* Example: $\frac{1}{2} = \frac{4}{8}$
 
#### Determining Proportion Equality
* To determine if a proportion is true, we use the **cross product** method, where each denominator is multiplied by the opposite fraction's numerator (diagonally across equal sign) & the determine if the resulting products are equal.
* Example: Determine equality of $\frac{1}{2} = \frac{4}{8}$
 * 8 * 1 = 8
 * 2 * 4 = 8
 
#### Solve Proportions
* When a proportion has a missing numerator or denominator (aka a variable), ex: $\frac{x}{63} = \frac{4}{7}$
* Solve by applying the cross product of one diagonal and then dividing the product by the remaining value.
* Example: Solve $\frac{x}{63} = \frac{4}{7}$
 * Multiply given diagnoal: 63 * 4 = 152
 * Divide product by remaining value: 252 $\div$ 7 = 36
 * Solution: $\frac{36}{63} = \frac{4}{7}$


## Percentages
* A ratio expressed as a fraction of 100 (i.e. the denominator is 100).
* A dimensionless number having no unit of measurment that can be applied to another number, called the **base**, to represent a proportion of that number.
* Denoted using a percent sign, %, and sometimes with the abbreviations 'pct' or 'pc'.

#### Converting Percentages to Fractions
* Convert percentages to fractions by putting the number before the percent sign as the numerator and 100 as the denominator, then simplify.
 * Example: $60% = \frac{60}{100} = \frac{3}{5}$
* Convert decimal percentages to fractions by putting the percent in the numerator over 100 and moving the decimal on top & bottom until the decimal is gone. Simplify.
 * Exmaple: $24.5% = \frac{24.5}{100} = \frac{245}{1000} = \frac{49}{200}$
* Convert mixed fraction percentages to a fraction by converting the numerator to an improper fraction & treating the larger fraction bar as division.
 * Example: $33\frac{1}{3}\% = \frac{\frac{100}{3}}{100} = \frac{100}{3} \div \frac{100}{1} = \frac{100}{3}(\frac{1}{100}) = \frac{100}{300} = \frac{1}{3}$

#### Converting Decimals to/from Percentages
* Convert percentages to decimals by moving the decimal to the left 2 places.
 * Example: 98% = 0.98
 * Example: 125% = 1.25
* Convert decimals to percentages by moving the decimal to the right 2 places.
 * Example: 0.17 = 17%
 
#### Converting Fractions to Percentages
* Convert fractions to percentages by dividing the numerator by the denominator and moving the decimal 2 places to the right.
 * Example: $\frac{3}{4} = 0.75 = 75\%$
 
## Calculating with Percentages
Easy way to remember which formula to use when calculating the base, amount, or pct: when finding the base, you divide by the percentage. When finding the percentage, you divide by the base. For the Amount, you multiply. See below for details.

#### Calculating Amounts: With Percentage & Base
* To find the numerical amount for a percentage of a given number.
* Calculate amount by converting the percentage to a decimal (move decimal 2 places to the left) and multiply it by the given base number.
* Example: 7% of 258 is = 0.07 * 258 = 18.06
 
#### Calculating Base: With Percentage & Amount
* To find the base number using the given the percentage & its equivalent amount.
* Calculate base by converting percenatage to a decimal (move decimal 2 places right) and dividing it into the given amount.
* Exmaple: 36 is 75% of = 36/.75 = 48

#### Calculating Percentage: With Amounts & Base
* To find the proportion of a given amount from a base number.
* Calculate the percentage by dividing the the given amoutn by the base.
* Example: percent of 36 is 9 = 9/36 = 1/4 = 0.25 = 25%

#### Percent Proportion
* A way to solve a percentage problem, where the percentage is written as a fraction in a proportion. One denominator will always be 100 and a variable represents the missing base, percent, or amount. Then solve as you would a regular proportion problem (See Proportions section above).
* Example: What's 45% of 80?
 * Rewrite as proportion: $\frac{45}{100} = \frac{x}{80}$
* Example: 125% is 64 of what number?
 * Rewrite as proportion: $\frac{125}{100} = \frac{64}{x}$
* Example: What percent of 27 is 9?
 * Rewrite as proportion: $\frac{x}{100} = \frac{9}{27}$


#### Calculating Percent Increases/Decreases
First find the difference between the new amount & the original amount. Then find the percentage of that to the original amount (Note that as shown above, to find the percent, you divide the amount by the base & convert to percent).
* Increase Example: What percent increase from 26 to 36?
 * Find difference: 36 - 26 = 10
 * Find percent of original (divide amt by base): 10/26 = .3846
 * Round & convert to percent: .385 = 38.5%
* Decrease Example: What percent decrease from 3.71 to 3.64?
 * Find difference: 3.71 - 3.64 = 0.07
 * Find percent of original (divide amt by base): 0.0188
 * Round & convert to percent: 0.019 = 1.9%

# Rational Numbers
* Rational numbers are a subset of real numbers that can be written as a ratio of 2 integers, i.e. can be represented fractionally in the form p/q, where p and q are integers, but q $\neq$ 0. 


### Integers
* All integers are rational numbers because they can be written as a fraction with a denominator of 1, e.g. $-8 = \frac{-8}{1}$
* Since all natural numbers and whole numbers are integers, they, too, are rational.

### Fractions
* All fractions are rational numbers, but not all rational numbers are fractions for the same reason all fractions are ratios but not vice versa. Rational numbers allow for negative integers in their terms. A fraction represents parts of a whole, while ratios represent a comparison.

### Some Decimals
* Since all fractions can be written as decimals, but not all decimals can be written as fractions, we can also talk about rational numbers in terms of decimals.
* There are 3 types of decimals: terminating(finite), recurring(infinitely repeating), and non-recurring(infinitely non-repeating). The first 2 are rational, while the last is not. 
 * But technically, all terminating decimals are also repeating since it repeats the digit 0 indefinitely. So, you could simply say repeating decimals are rational, while non-repeating decimals are not.
 * Additionally, whether a rational number terminates or repeats depends on the base used. In base 10, $\frac{1}{20}$ terminates and $\frac{1}{7}$ repeats. In base 2, both repeat. In base 14, $\frac{1}{7}$  terminates and $\frac{1}{20}$  repeats. In base 70, both terminate.

# Irrational Numbers
* Irrational numbers are a subset of real numbers that cannot be written as a ratio of 2 integers.
* Discovered by Hippasus of Metapontum in the 5th century BCE while trying to find the hypotenuse of an isosceles right triangle whose base sides are 1, giving it a hypotenuse of $\sqrt{2}$ (a non-terminating non-repeating decimal), and producing a (most likely geometrical) proof of the irrationality of the square root of 2 while trying to represent it as a fraction.
 * Legend says that Pythagoras believed in the absoluteness of numbers & couldn't accept the existence of irrational numbers, but couldn't disprove it. So he sentenced Hippasus to death by throwing him into the sea.

### Non-Recurring Decimals
* Of the 3 types of decimals, only the infinite, non-repeating decimals are irrational.
* E.g. pi goes on forever with no pattern: $\pi = 3.141592...$

### Some Square Roots
* Square roots that aren't perfect squares are irrational. 
 * A square root that is a perfect square will always be a whole number, e.g. $\sqrt{9} = 3$
 * A square root that is not a perfect square will always be a non-recurring decimal, e.g. $\sqrt{5} = 2.2360679...$

# Real Numbers
* Includes all rational and irrational numbers.
* A real number is a value of a continuous quantity that can represent a distance along a line (or alternatively, a quantity that can be represented as an infinite decimal expansion). 
 * Can also measure quantities such as time, mass, energy, velocity, etc.
* The adjective real in this context was introduced in the 17th century by René Descartes, who distinguished between real and imaginary roots of polynomials.


## Properties of Real Numbers

#### Commutative Properties
* Changing the order of terms between multiplication or between addition doesn't change the result.
* The sign always stays with the term.
 * The Commutative Property does not apply to subtraction or division. However, since the term keeps the sign, subtraction is acceptable if you make sure to keep the sign with the term.
* **Commutative Property of Addition**: If a and b are real numbers, then a + b = b + a
* **Commutative Property of Multiplication**: If a and b are real numbers, then ab = ba
* Example: 2 + (-5) = -5 + 2 = -3
* Example: 2 - 5 = -5 + 2 = -3

#### Associative Properties
* Changing the order of operations between multiplication or between addition doesn't change the result, i.e. rearranging grouping symbols to combine differently.
 * Does not apply to subtraction or division.
* **Associative Property of Addition**: If a,b & c are real numbers, then (a + b) + c = a + (b + c)
* **Associative Property of Multiplication**: If a,b & c are real numbers, then (ab)c = a(bc)
* Example: $\frac{3}{8}(17 * \frac{8}{3}) = (\frac{3}{8} * \frac{8}{3})17 = 1 * 17 = 1$

#### Distributive Property
* Using a multiplier on an expression of addition causes each term between addition to be multiplied.
* If a, b, & c are real numbers, then a(b + c) = ab + ac
* Example: If 3 friends want movie tickets each at \\$9.25$, you can get the total by multiplying 9 by 3, then multiplying .25 by 3, and adding them together.

#### Identity Properties
* **Identity Property of Addition**: Adding zero to any value doesn't change the value.  
 * For any real number a, a + 0 = a
 * 0 is called the "additive identity".
* **Identity Property of Multiplication**: Multiplying any number by 1 doesn't change the value. 
 * For any real number a, a1 = a
 * 1 is called the "multiplicative identity". 

#### Inverse Properties
* These are essentially the opposite of the Identity Properties.
* **Inverse Property of Addition**: Adding the opposite signed number to the number results in 0.
 * For any real number a, a + (-a) = 0
 * -a is the "additive inverse" of a.
* **Inverse Property of Multiplication**: Any number multiplied by its reciprocal results in 1.
 * For any real number a $\neq$0, a($\frac{1}{a}$) = 1
 * A number's reciprocal is it's "multiplicative inverse", such that the number isn't 0.
 
#### Properties of Zero
* Zero is even.
* Zero is neither negative nor positive (the only number with this property).
* Zero is an integer.
* Zero is a multiple of all numbers.
* **Zero Property of Multiplication**: Any number multiplied by 0 results in 0.
 * For any real number a, a0 = 0
* **Zero Property of Division**: 0 divided by any number is 0. 
 * For any real number a$\neq$0, $\frac{0}{a}$ = 0
 * Dividing by zero is undefined ($\frac{a}{0}$ = undefined)
 
#### Properties of Equality
* These are used to solve equations, by doing the same operation on both sides of an equation to isolate a variable.
* **Subtraction Property of Equality**: For any real number a,b,c, if a = b, then a-c = b-c
* **Addition Property of Equality**: For any real number a,b,c, if a = b, then a+c = b+c
* **Division Property of Equality**: For any real number a,b,c, & c $\neq$ 0, if a = b, then $\frac{a}{c} = \frac{b}{c}$
* **Addition Property of Equality**: For any real number a,b,c, if a = b, then ac = bc

# Algebraic Expression
* Mathematical phrase that contains a group of terms (numbers, variables, or both) each separated by an operator.
* Algebraic Expressions must have at least one variable.
 
### Parts of an Expression
* **Variable** is a symbol for a number we don't know yet. It is usually a letter like x or y. They have a context & it assumed that it's reset for every new problem.
* **Constant** is a number on its own whose value does not change.
* **Coefficient** is a number used to multiply a variable (4x means 4 times x, so 4 is a coefficient)
 * Variables on their own (without a number next to them) have an implied coefficient of 1 (x is really 1x)
 * Sometimes a coefficient is a letter like a or b instead of a number, Example: ax2 + bx + 4
    * x is a variable
    * a and b are coefficients
    * 4 is a constant
* **Operator** is a symbol (such as +, -, etc) in algebraic expressions that shows an operation, usually + or -, but never multiplication.
 * Arithmetic Operators are performed on numbers whereas algebraic operators can be performed on variables, algebraic expressions, and more generally on elements of algebraic structures and are formatted differently.
   * Multiplication is generally represented by coefficients or grouping symbols in lieu of the * operator.
   * Division is generally represented using fractions in lieu of the $\div$ operator.
   * Exponents may have a variable in the base or power
   * Roots may have a variable
* **Term** is a constant, a variable, a coefficient, or numbers and variables multiplied together and are separated by an operator.
* **Grouping Symbols** usually brackets [ ], parenthesis (), braces {} used to determine which terms should be evaluated together. These grouping symbols are also used for multiplication when no oporator is present between terms.
 * Other symbols used as grouping symbols are absolute value symbols |x|, the fractional bar $\frac{n}{d}$, or the radical symbol $\sqrt{}$.

# Evaluating Algebraic Expressions
* Evaluating algebraic expressions employs the use of logical rules that apply to all real numbers so as to simplify an expression.
* We solve an expression using the Order of Operations & evaluate it by using substitution.

#### Order of Operations
* Order of Operations are the rules denoting the order in which terms in an expression are evaluated. BODMAS (or PEMDAS) is an acronym helpful for memorizing the order:
 1. **B**rackets first: Evaluate terms within grouping symbols (incl fraction bar, abs val, radical)
    1. For brackets within brackets, work inner to outer.
 2. **O**rders (i.e. Exponents and Roots, etc.)
    1. Exponents of Exponents are evaluated top down.
 3. **D**ivision and **M**ultiplication (left-to-right)
 4. **A**ddition and **S**ubtraction (left-to-right)
* But for multiplication & addition, we can use the Commutative & Associative Properties to rearrange order or operations in a way that could reduce work.

#### Substitution
* Substitution is plugging in numbers where the letters are, if you have their values.
* When substituting negative numbers, wrap negative numbers in parenthesis.
* To reduce work, simplify the expression as much as possible before plugging in values.

## Adding/Subtracting Monomials
* We can add or subtract monomials if they are like terms.
* Terms are called **Like Terms** if they have the same variables and exponents. All constant terms are also like terms. Radicals must have the same index AND radicand. Else, they're **Unlike Terms**.
* If a term has multiple variables or coefficients, the order of the variables/coefficients that aren't next to the exponent does not matter and they are still considered like terms, as per the Associative Property of Multiplication.
 * Example: 3a$^2$b + ba$^2$ = 4a$^2$b
 
#### Combine Like Terms 
* Like terms can be added together or subtracted from one another simply by adding/subtracting the coefficients.
* This is essentially using the Distributive Property in reverse (factoring), but without the added step.
 * Example: 5x + 3x = (5 + 3)x = 8x
* A variable without a number coefficient still has an implied coefficient of 1.
* If an expression has a lot of terms, it's best to rearrange the expression with like terms next to each other before combining. This rearrangement is per The Commutative Property of Addition, where terms can be rearranged without changing the values.
 * Example: 2x + 3 + a$^2$ + 4x - 1
  * Rearrange: a$^2$ + 2x + 4x + 3 - 1
  * Simplify: a$^2$ + 6x + 2


## Multiplying Monomials
* Multiply constants only with coefficients
 * Example: 3x * 2 = 6x
 * Example: $3x^3 * 2 = 6x^3$
* Multiply like variables by raising them to the power of factors multiplied.
 * Example: x * x = $x^2$
 * Example: x * $x^2 = x^3$
* Multiply like base terms with coefficients by multiplying coefficients & raising variables to power of factors.
 * Example: 3x * 4x = 12x$^2$
* Multiply unlike base terms with coefficients by multiplying coefficients & leaving bases unchanged.
 * Example: 2a * 3b = 6ab
* Can multiply terms with exponents if the variable is the same (exponents can be different) by multiplying the coefficients & adding the exponents. Note: More exponent examples in 'Exponents' section.
 * Example: $(4x^2)(-5x^3) = -20x^5$
 * Example: $3x^3 2x^4 = 6x^7$
 
#### Multiplying Exponents
* When multiplying exponents with the same base, add the exponents for the result, as per the 'Product Property of Exponents'.
 * Example: $x^2 * x^3 = x^5$
* Parenthesis can be further simplified by applying the exponent to each base, as per the 'Product to a Power Property'.
 * Example: The difference between $2x^2$ and $(2x)^2$
   * $2x^2 = 2 * x * x$
   * $(2x)^2 = 2x * 2x$ OR $2^2 x^2 = 4x^2$
 * Example: More than 2 bases
   * $(3xy)^3 = 3^3x^3y^3 = 27x^3y^3$
 * Example: Raising to the zeroith power will always be 1.
   * $(2x)^0 = 1$
   * $(-7xy)^0 = 1$
* When raising a power to a power, multiply the exponents, as per the 'Power Property of Exponents'.
 * Example: Simplify powers of powers using Power Property & further simplify using Product Property.
   * $(x^2)^6(x^5)^4 = x^{12}x^{20} = x^{32}$
 * Example: Simplify using Product to a Power Prop & further simplify using Power Prop.
   * $(-7x^3y^4)^2 = (-7)^2(x^3)^2(y^4)^2 = 49x^6y^8$
* Use Communtative Property to simplify long exponent results.
 * Example: Simplify & rearrange like terms to further simplify.
   * $(6n)^2(4n^3) = 6^2 n^2 4n^3 = 36 * 4 * n^3 * n^2 = 144n^5$

## Dividing Monomials (Fractions)
* Work the same way as regular fractions & operations with variables
 * Simplification Example: $\frac{5xy}{15x} = \frac{5y}{15} = \frac{y}{3}$
 * Addition Exmaple: $\frac{x}{3} + \frac{2}{3} = \frac{x + 2}{3}$
 * Multiplication Example: $\frac{12}{5}(-20x) = \frac{12}{5}(-\frac{20x}{1}) = -\frac{240x}{5} = -48x$
* Another strategey is "clearing the equation of fractions" whereby both sides of equation are multiplied by the LCD of all the fractions. Distribute existing parens first, if applicable.
 * Example: $\frac{1}{8}x + \frac{1}{2} = \frac{1}{4}$
   * Find LCD of all fractions: 8
   * Multiply both sides: $\frac{8}{1}(\frac{1}{8}x + \frac{1}{2}) = \frac{8}{1}(\frac{1}{4})$
   * Evaluate: x + 4 = 2
   * Solve: x = -2
* Can also "clear the equation of decimals" the same way. Multiply both sides by the LCD, which is the digit place of the decimal. Distribute existing parens first, if applicable.
 * Example: 0.8x - 5 = 7
   * Find LCD: 10
   * Multiply both sides: 10(0.8x - 5) = 10(7)
   * Evaluate with distribution: 8x - 50 = 70
   * Solve: x = 15
* Complex fractions that can be simplified in the numerator &/or denominator
 * The fraction bar works as a grouping symbol so the numerator & denominator must be evaluated before dividing the numerator by the denominator.
 * Example: Evaluate $\frac{(\frac{1}{2})^2}{4+3^2}$
   * $\frac{\frac{1}{4}}{13} = \frac{1}{4} * \frac{1}{13} = \frac{1}{52}$

#### Dividing Exponents
* When dividing exponents with the same base & the denominator exponent is smaller, subtract from numerator exponent. If numerator exponent is smaller, subtract from denominator exponent & put over 1, as per the 'Quotient Property of Exponenets'. Essentially, if numerator is larger, result is $\frac{x}{1} = x$, else result is $\frac{1}{x}$
 * Example: Numerator larger
   * $\frac{x^5}{x^2} = x^3$
 * Example: Numerator smaller
   * $\frac{x^2}{x^5} = \frac{1}{x^3}$
 * Example: Different bases have different exponent ratios, determining which part of fraction they'll be in.
   * $\frac{24a^5b^3}{48ab^4} = \frac{1a^{5-1}}{2b^{4-3}} = \frac{a^4}{2b}$
* If exponents are the same, the result is always 1.
 * Example: Exponents are the same
   * $\frac{x^3}{x^3} = x^{3-3} = x^0 = 1$
* When raising a fraction to a power, apply the exponent to both numerator & denominator as per the 'Quotient to a Power Property'. Reduce as much as possible inside parenthesis first to reduce unnecessary steps.
 * Example: Can't further reduce inside parens
   * $(\frac{-5}{m})^3 = \frac{(-5)^3}{m^3} = \frac{-125}{m^3}$
 * Example: Reduce inside parens
   * $(\frac{x^7}{x^3})^2 = (x^4)^2 = x^8$
* The fraction bar is a grouping symbol so evaluate top & bottom first.
 * Example: $\frac{(x^2)^3}{x^5} = \frac{x^6}{x^5} = x^{6-5} = x$
* Negative exponents represent a less simplified form of the 'Quotient Property of Exponents' where the numerator is smaller. Simplify expressions by removing the negative exponents.
 * Example: $x^{-4}x^6 = (\frac{1}{x^4})(\frac{x^6}{1}) = \frac{x^6}{x^4} = x^2$
   
   
## Multiplying Polynomials (Distribution)
* We use distribution to get rid of parenthesis surrounding binomials when simplifying expressions.
* Algebraic multiplication on a binomial can occur using the Distributive Property with monomials or other binomial, trinomial, or polynomial expressions. 

#### Multiply Monomials with Binomials 
* Multiplying each of the terms in the binomial within the parentheses by another term (monomial) outside the parentheses.
 * Example: 3(x + 2) = 3x + 6
 * Example: 4(x - 8)-(x + 3) = 4x - 32 - x - 3 = 3x - 35

#### Multiply Binomials with Binomials
* Multiplying by another expression (binomial) in parenthesis.
* **FOIL** (the acronym for first, outer, inner and last) method is an efficient way of remembering how to multiply two binomials in a very organized manner. In FOIL, "First" means to multiply the terms which occur in the first position of each binomial. "Outer" means to multiply the first term in the first binomial with the last term in the last binomial. Etc.
 * Example: (2x + 9)(3x + 4)
   * First: 2x * 3x, Outer: 2x * 4, Inner: 9 * 3x, Last: 9 * 4
   * Evaluate: $6x^2 + 8x + 27x + 36$
   * Simplify: $6x^2 + 35x + 36$
   
#### Multiplying Trinomials+ 
* The FOIL Method only works binomial to binomial, but the vertical method or distribution works for everything.
* Multiply each term with each other term.


## Factoring Polynomials
* Used to find the 'Greatest Common Factor' (GCF) of two or more monomials or to factor out an expression, like doing distribution in reverse.

#### Find the GCF of Two Monomials
* Find the largest expression that is a factor of all the expressions.
* Done similarly to factoring numbers, except it's with expressions.
* To factor a monomial, factor the coefficients into primes & expand all the exponents into single variables.
* Then find the GCF by lining the factors into columns, keep the ones that match each monomial, then simplify.
 * Example: Find GCF $12x^2$ and $18x^3$
   * T1: $12x^2$ = 2 * 2 * 3 * .... x * x ....
   * T2: $18x^3$ = 2 * .... 3 * 3 * x * x * x
   * Common = 2 * .... 3 * .... x * x ....
   * GCF = $6x^2$

#### Find the GCF of a Polynomial Expression
1. Find the GCF between each term in the expression as done above.
 1. If the leading coefficient is negative, turn the GCF negative.
2. Divide each term by the GCF.
3. Put the divided terms in parenthesis with the GCF on the outside.

> Example: Factor $3x^2 + 6x + 9$
 * GCF: 3
 * Divide each by 3: $x^2, 2x, 3$
 * Multiply by GCF: $3(x^2 + 2x + 3)$

## Radicals
* Evaluated first as they're considered grouping symbols in the Order of Ops.
* Can evaluate radicals by thinking of an expression whose square would equal the radicand.
 * Example:

\sqrt{9x^2} &= ? \\
(?)^2 &= 9x^2 \\
(3x)^2 &= 9x^2 \quad\text{(thus...)} \\
\sqrt{9x^2} &= 3x
\end{align*}$

* Can also evaluate by thinking of the radical canceling out the square exponent, but still needing to find the square root of the coefficient.
 * Example: $\sqrt{36x^2y^2} = 6xy$

* Equations that use the equal symbol are called Equalities whereas equations with any of the inequality symbols are called Inequalities and are evaluated using a slightly different process.
* Always simplify the equation before trying find a solution (combine like terms, distribute parens).
* Equations are solved by isolating the variable. 
 * If multiple variables, put variables on one side.

### Evaluating Equalities
* Equalities are equations with a equals sign (=)
* Solve for a variable in an equation by using the Addition/Subtraction/Division/Multiplication Properties of Equalities to isolate the variable. Properties are executed using the Order of Operations in reverse.  
 * **Example:** Solve for **x** in the equality **2x + 4 = 16**

> $\begin{align*}
2x + 4 &= 16 \\
2x &= 12 \quad\text{(minus 4 from both sides)} \\
x &= 6 \quad\text{(divide by 2 on both sides)}
\end{align*}$

* Solve fractional coefficients by either dividing both sides by the fraction OR multiplying both sides by the reciprocal.
 * Multiplying a fraction by it's reciprocal = 1
 * **Example:** Solve for x in the equation $\frac{3}{4}x = 24$
 
> $\begin{align*}
\frac{3}{4}x &= 24 \\
\left(\frac{4}{3}\right)\frac{3}{4}x &= \left(\frac{4}{3}\right)24 \quad\text{(multiply by the reciprocal)} \\
1x &= \frac{96}{3} \quad\text{(simplify)} \\
x &= 32 \quad\text{(simplify)}
\end{align*}$

* Solve negative coefficients by dividing by -1
 * **Example:** Solve for x in the equation -x = 15

> $\begin{align*}
-x &= 15 \\
x &= -15
\end{align*}$

* Solve by keeping order of negatives when using Subtraction Property of Equality.
 * **Example:** Solve for a in the equation P = a + b + c

> $\begin{align*}
P &= a + b + c \\
P - b - c &= a
\end{align*}$

* Solve using Division Property of Equality.
 * **Example:** Solve for P in the equation I = Prt

> $\begin{align*}
I &= Prt \\
\frac{I}{rt} &= P
\end{align*}$

 * **Example:** Solve for y in the equation 3x + 2y = 18

> $\begin{align*}
3x + 2y &= 18 \\
2y &= 18 - 3x \\
y &= \frac{18 - 3x}{2}
\end{align*}$

### Evaluating Inequalities
* Equations with a less than(<), greater than(>), greater than or equal to($\ge$), or less than or equal to($\le$) sign
* Result is any number larger/smaller than specific value
* Similar to solving for Equalities except the relationship inequality sign flips when:
 * 1. Dividing/multiplying by a negative number.
 * 2. Sides are swaped.
 * **Example:** Solve for x in the inequality -3x + 5 > 29

> $\begin{align*}
-3x + 5 &> 29 \\
-3x &> 24 \quad\text{(minus 5 from both sides)} \\
x &> -8 \quad\text{(divide by -3 on both sides)} \\
x &< -8 \quad\text{(flip the sign)}
\end{align*}$


### System of Equations
* Puzzle with multiple variables that can be solved if at least as many unique equations exist as variables in the system of equations.
* Solve for one variable first, plug it in to the other equation & solve for the other var, then plug back into 1st equation
 * **Example:** Solve for **x** & **y** in the equations **2x + 2y = 14** and **x - y = 3**.

> EQUATIONS: <br>
$\begin{equation*}
2x + 2y = 14 \quad\text{(Equation 1)} \\
x - y = 3 \quad\text{(Equation 2)} 
\end{equation*}$

> PROCESS: <br>
$\begin{align*}
x - y &= 3 \quad\text{(Equation 2: solve for x)} \\
x &= y + 3 \quad\text{(add y to both sides)} \\
\\
2x + 2y &= 14 \quad\text{(Equation 1: substitue x for E2 value & solve for y)} \\
2(y+3) + 2y &= 14 \quad\text{(replace x with y+3)} \\
2y + 6 + 2y &= 14 \quad\text{(simplify with distribution)} \\
4y + 6 &= 14 \quad\text{(add like terms)} \\
4y &= 8 \quad\text{(minus 6 from both sides)} \\
y &= 2 \quad\text{divide by 4 on both sides} \\
\\
x &= y + 3 \quad\text{(Equation 2: substitue y for E1 value & solve for x)} \\
x &= (2) + 3 \quad\text{(plug in 2 for y)} \\
x &= 5 \quad\text{(simplify)}
\end{align*}$

> SOLUTION: <br> x = 5 <br> y = 2


# Cartesian Coordinate System
* Developed by Rene Descartes (thus Cartesian), it revolutionized mathematics by providing the first systematic link between Euclidean geometry and algebra. Using the Cartesian coordinate system, geometric shapes (such as curves) can be described by Cartesian equations: algebraic equations involving the coordinates of the points lying on the shape.
* Rectangular grid system that uses the horizontal number line, the **x-axis**, with a vertical number line, the **y-axis**, crossing perpendicularly at 0.
 * Aka the x-y plane, the coordinate plane, or the Cartesian plane.
* Every point is represented by an **ordered pair** in parenthesis, the first number being the **x-coordinate** of the point, and the second number being the **y-coordinate**, i.e. (x,y).
* The point at which the x & y axes intersect at 0 is called the **origin** & has the coordinates (0,0).
 * Points with a y-coordinate equal to 0 are on the x-axis, and have coordinates (a,0).
 * Points with an x-coordinate equal to 0 are on the y-axis, and have coordinates (0,b).
* To distinguish between two or more points, we use **subscripts**, which is a small number written to the right of, and a little lower than, a variable like $(x_1,y_1)$, read 'x sub 1, y sub 1' and $(x_2,y_2)$, read 'x sub 2, y sub 2', and so on.
* The x- and y-axes divide the plane into 4 areas, called **quadrants**. The quadrants are identified by Roman numerals, beginning on the upper right and proceeding counterclockwise.
 * There is a pattern of signs in the coordinates for each quadrant.
   * Quadrant I: (+,+)
   * Quadrant II: (-,+)
   * Quadrant III: (-,-)
   * Quadrant IV: (+,-)

## Terms
* **Abscissa**: The x -coordinate, the distance from a point to the vertical or y-axis, measured parallel to the horizontal or x -axis.
* **Ordinate**: The y-coordinate, represents the directed distance of the point from the x-axis.
* **Point Plotting**: To place a point on a coordinate plane by using its x and y coordinates.
* **Graphing**: Plotting the collection of all points whose coordinates satisfy a given relation (such as a function).
* **Function**: Subset of equations that have solutions based on the input for the values of the variables.
* **Graph of a Function**: the graph of a function $f$ is the set of ordered pairs $(x,y)$, where $f(x)=y$, thus satisfying a given relation, and will cross the vertical line at two or more points.
 * Functions cannot always be graphed.
 * In the common case where $x$ and $f(x)$ are real numbers, these pairs are Cartesian coordinates of points in two-dimensional space and thus form a subset of this plane.
* **Graph of an Equation**: The graph of an equation drawn using a single line for linear and parabola for higher-degree equations should only intersect at one point with a vertical line drawn in the graph.
 * Equations can always be graphed because of the definite value of x.
* **Parallel Lines**: two lines that never intersect & have the same slope. Denoted by comparisons:
 * is parallel with: $\|$
 * is NOT paralell with: $\nparallel$
* **Perpendicular Lines**: two lines that intersect at a 90$^{\circ}$ angle. Denoted by comparisons:
 * is perpendicular to: $\perp$
 * is NOT perpendicular to: $\not\perp$



# Linear Equations
* Aka 'One-degree equation', highest power of the variable is 1, which is typically not shown, but implied.
* Linear relationships can be expressed using a mathematical equation or graphical format.
* The graph of a linear equation will always be a straight line that represents all the solutions to the equation, since any and every point on the line is an ordered pair (x,y) solution to the equation.

#### Linear Equation in One Variable
* Standard Form: ax + b = 0, where a,b are real numbers and a$\neq$0 
 * x is the "one variable", which means there is only 1 solution in the form of $x = -\frac{b}{a}$
 * The equation is **constant** since the only value never changes.
* The graph of linear equations in one variable will always be either a horizontal or vertical line, depending on which axis variable is given in the equation. For any ordered pair (x,y) used to graph the line, the constant variable will always be the same value, while the other varaible can be any & all values along the constant's axis. Thus, x & y are independent of each other.

#### Linear Equation in Two Variables
* Standard Form: ax + by + c = 0, where a,b,c are real numbers and a,b are not both 0 
 * x and y are the "two variables", which mean there are infinitely many solutions. & their values are **dependent** on each other.
 * For every number that is substituted for x, there is a corresponding y value. 
* The graph of a linear equation in two variables will always be a slanted line that passes through both axes. Any ordered pair solutions to the equation are points that lie along the same line. Thus, x & y are dependent on each other.

## Intercepts
* The points at which a line intersect the x- & y-axes.
 * The **x-intercept** is the point, (a,0), where the graph crosses the x-axis. The x-intercept occurs when y is zero.
 * The **y-intercept** is the point, (0,b), where the graph crosses the y-axis. The y-intercept occurs when x is zero.
* Can graph a line using the intercepts from its equation.

#### Calculating Intercepts with Equation
* We can find the x & y intercepts by using a line's equation & this is often the easiest way to graph a linear equation.
* Since the x-intercept occurs when y is zero and the y-intercept occurs when x is zero gives, to find the x-intercept, let y=0 and solve for x. To find the y-intercept, let x=0 and solve for y.
* Example: Find the intercepts of 2x + y = 6
 * Y-Intercept:
   * x = 0
   * 2(0) + y = 6
   * y = 6
   * (0,6)
 * X-Intercept: 
   * y = 0
   * 2x + 0 = 6
   * x = 3
   * (3,0)


## Slope
* Visualy, slope is the steepness of a line.
* Conceptually, slope is the ratio of vertical change to horizontal change between two points on the plane or a line. So, slope equals the ratio of the rise over the run, measuring the rate of change for a line.
* **Rise**: Vertical change of a line (y-change).
* **Run**: Horizontal change of a line (x-change).
* **m**: Variable that's always used to represent slope.
* Formula: $m = \frac{rise}{run}$

* A line can slant upwards or downwards, but it is always read from left to right. Therefore, the rise can be positive or negative, but the run will always be positive. Thus, the rise determines the sign of a slope.
 * If a line slants upward, the slope is positive. 
 * If a line slants downward, the slope is negative.
* Slope is the hypotenuse of a right triangle, where the height & length of the sides are the rise & run, respectively. On a grid, if a line is vertical or horizontal, a right triangle cannot be formed because either the rise (height) or the run (length) is 0.
 * Horizontal lines have a slope of 0 because the rise is 0 and $\frac{0}{x} = 0$ (0 divided by anything is 0).
 * Vertical lines have a slope of 'undefined' because the run is 0 and $\frac{y}{0} = undefined$ (you can't divide by 0).

#### Determining Slope/Points/Lines Visually
* The slope of a line on a grid can be determined by counting the rise & run from one point to the other.
 * Count the Rise: Start from the leftmost point and count the intervals vertically to the second point.
  * If you count upward, the rise is positive.
  * If you count downward, the ries is negative.
 * Count the Run: From the point you're at, count the intervals horizontally to the second point.
 * Put the rise over run in a fraction.
* Likewise, a point on grid can be determined by counting in the same way using slope from another point. 
 * For graphs with larger numbers, it's helpful to use a larger equivalent fraction of the slope for accuracy.

#### Slope Formula
* If you have the coordinates of two points, you can get the slope by subtracting the first point from the second point to get the rate of change. 
* Formula: $m = \frac{y_2 - y_1}{x_2 - x_1}$
* You will get the same answer regardless of which point is subtracted from which.
* Example: Find the slope of (1, 2) and (3, 5)
 * $m = \frac{5-2}{3-1}$
 * $m = \frac{3}{2}$
* The slope formula is the same as $m = \frac{rise}{run}$ because rise represents a 'change in y' and run represents a 'change in x'.

#### Calculating Points Given a Point & Slope
* If you have the coordinates of 1 point and the slope of a line, you can calculate other points on the line to graph, since y is associated with the rise & x with the run.
* Unofficial Formula: $(x_1,y_1) = (x + run, y + rise)$
* To get the second point for a graph, you add the slope to the given point by adding the rise to y and the run to x. You can add the slope again to the result for a third point.
* Example: For a point, (3, 2), and a slope of 2, find another point.
 * $(x_1,y_1) = (3 + 1, 2 + 2)$
 * $(x_1,y_1) = (4,4)$
 * $(x_2,y_2) = (4 + 1, 4 + 2)$
 * $(x_2,y_2) = (5,6)$
 
#### Slope-Intercept Form
* Used to easily determine slope & y-intercept.
* The standard form of a linear equation in two variables is ax + by + c = 0, where x & y are the variables. But a linear equation in two variables can be written in different forms as long as there are two variables, ex: y = ax + b.
* Slope-intercept form, or "y-form", is a linear equation in two variables that's formatted as y = mx + b
 * When a linear equation is solved for y, the coefficient of the x term is the slope, represented as m rather than 'a', and the constant term, b, is the y-coordinate of the y-intercept. 
* To convert an equation to slope-intercept form, simply solve for y.
* Any equation formatted in this way automatically gives you a point & the slope.
* Example: For the equation $y = \frac{1}{2} + 3$
 * $m = \frac{1}{2}$
 * y-intercept = (0,3)

#### Point-Slope Formula
* Used to convert points into an equation.
* If you have the coordinates of 1 point and the slope of a line, you can calculate the equation of a line.
* Formula: $y - y_1 = m(x - x_1)$
* Example: Find the equation for a line with a point (3, 2) and a slope of 2.
 * y - 2 = 2(x - 3)
 * y - 2 = 2x - 6
 * y = 2x - 6 + 2
 * y = 2x - 4
* Since the equation is in slope-intercept form, we have another point, the y-intercept (0,4), to graph a line.


## Graphing Linear Equations
* The solutions for linear equations are in ordered pair form (x,y), which are used to graph a line of the equation.
* Use an optimal method to find 3 points depeding on the equation 
 * Solve for at least 3 points to ensure accuracy.
* Point-plot the ordered pair solutions in the table.
* Draw a line through the points, extending the line to fill the grid &/or drawing arrows on both sides of the line to show indefinte extention.
* If the line isn't straight, something went wrong, & you need to recheck calculations or replot the points.

#### Choosing an optimal method for getting points
* There are many ways to get point coordinates for a linear equation, the most optimal of which depends on the format &/or variables of an equation, or its slope.
 * If the equation has only one variable, the value is constant, & it is a vertical or horizontal line.
 * If both x & y are on the same side of the equation (of the form ax + by = c), use the intercepts & sub x for any number for the 3rd point.
   * Note that fractions are more difficult to graph because they require more precision. If an equation has both variable on one side, isolate the variable that doesn't have a coefficient to avoid getting fractional cooridinates.
 * If y is isolated on one side of the equation, use slope-intercept form.
 * If the slope & 1 point are given instead of an equation, use the slope method.


## Interpreting Real World Applications of Slope-Intercept
* Many real-world applications are modeled by linear equations. 
* Usually when a linear equation models a real-world situation, different letters are used for the variables, instead of x and y. The variable names remind us of what quantities are being measured.

#### Temperature Conversion
* $F = \frac{9}{5}C + 32$, where F is Fahrenheit scale and C is the Celsius scale.
* Convert Celsius temperatures to Fahrenheit by substituting a C value & solving for F.
* $m = \frac{9}{5}$, means the temp Fahrenheit (F) increases 9 degrees when the temp Celsius (C) increases 5 degrees.
* F-intercept = 32, means that when the temperature is 0° on the Celsius scale, it is 32° Fahrenheit.

#### Female Height by Shoe Size
* h = 2s + 50, where h is height and s is shoe size.
* Estimates female height based on shoe size.
* m = 2, means that the height, increases by 2 inches when the shoe size increases by 1.
* h-intercept = 50, means that when the shoe size is 0, the height is 50 inches.

#### Temperature by Cricket Chirps
* $T = \frac{1}{4}n + 40$, where T is temperature in degrees Fahrenheit and n is number of cricket chirps per minute.
* $m = \frac{1}{4}$, means that the temp increases 1 degree when the number of chirps increases by 4 per minute.
* T-intercept = 40, means that when the number of chirps is 0, the temperature is 40° Fahrenheit.

#### Custom Example

 * The fixed cost is always the same regardless of how many units are produced. This is the cost of rent, insurance, equipment, advertising, and other items that must be paid regularly. 
 * The variable cost depends on the number of units produced. It is for the material and labour needed to produce each item.
* Example: Stella running a pizza business needs to determine the relationship between zir weekly cost and pizzas ze sells.
* C = 4p + 25, where C is the weekly cost, in dollars, and p is the number of pizzas sold.
* m = 4, means the cost increases by \$4 for each pizza sold.
* C-intercept = 25, is the fixed cost that exists even if ze sells no pizzas.
