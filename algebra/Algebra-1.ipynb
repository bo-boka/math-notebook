# Cartesian Coordinate System
* Developed by Rene Descartes (thus Cartesian), it revolutionized mathematics by providing the first systematic link between Euclidean geometry and algebra. Using the Cartesian coordinate system, geometric shapes (such as curves) can be described by Cartesian equations: algebraic equations involving the coordinates of the points lying on the shape.
* Rectangular grid system that uses the horizontal number line, the **x-axis**, with a vertical number line, the **y-axis**, crossing perpendicularly at 0.
 * Aka the x-y plane, the coordinate plane, or the Cartesian plane.
* Every point is represented by an **ordered pair** in parenthesis, the first number being the **x-coordinate** of the point, and the second number being the **y-coordinate**, i.e. (x,y). The phrase "ordered pair" implies that the order is important.
* The point at which the x & y axes intersect at 0 is called the **origin** & has the coordinates (0,0).
 * Points with a y-coordinate equal to 0 are on the x-axis, and have coordinates (a,0).
 * Points with an x-coordinate equal to 0 are on the y-axis, and have coordinates (0,b).
* To distinguish between two or more points, we use **subscripts**, which is a small number written to the right of, and a little lower than, a variable like $(x_1,y_1)$, read 'x sub 1, y sub 1' and $(x_2,y_2)$, read 'x sub 2, y sub 2', and so on.
* The x- and y-axes divide the plane into 4 areas, called **quadrants**. The quadrants are identified by Roman numerals, beginning on the upper right and proceeding counterclockwise.
 * There is a pattern of signs in the coordinates for each quadrant.
   * Quadrant I: (+,+)
   * Quadrant II: (-,+)
   * Quadrant III: (-,-)
   * Quadrant IV: (+,-)

## Terms
* **Abscissa**: The x -coordinate, the distance from a point to the vertical or y-axis, measured parallel to the horizontal or x -axis.
* **Ordinate**: The y-coordinate, represents the directed distance of the point from the x-axis.
* **Point Plotting**: To place a point on a coordinate plane by using its x and y coordinates.
* **Graphing**: Plotting the collection of all points whose coordinates satisfy a given relation (such as a function).
* **Function**: Subset of equations that have solutions based on the input for the values of the variables.
* **Graph of a Function**: the graph of a function $f$ is the set of ordered pairs $(x,y)$, where $f(x)=y$, thus satisfying a given relation, and will cross the vertical line at two or more points.
 * Functions cannot always be graphed.
 * In the common case where $x$ and $f(x)$ are real numbers, these pairs are Cartesian coordinates of points in two-dimensional space and thus form a subset of this plane.
* **Graph of an Equation**: The graph of an equation drawn using a single line for linear and parabola for higher-degree equations should only intersect at one point with a vertical line drawn in the graph.
 * Equations can always be graphed because of the definite value of x.



# Linear Equations
* Aka 'One-degree equation', highest power of the variable is 1, which is typically not shown, but implied.
* Linear relationships can be expressed using a mathematical equation or graphical format.
* The graph of a linear equation will always be a straight line that represents all the solutions to the equation, since any and every point on the line is an ordered pair (x,y) solution to the equation.

#### Linear Equation in One Variable
* Standard Form: ax + b = 0, where a,b are real numbers and a$\neq$0 
 * x is the "one variable", which means there is only 1 solution in the form of $x = -\frac{b}{a}$
 * The equation is **constant** since the only value never changes.
* Can be represented on a number line.
* In an x-y coordinate system, the graph of linear equations in1V will always be either a horizontal or vertical line, depending on which axis variable is given in the equation. For any ordered pair (x,y) used to graph the line, the constant variable will always be the same value, while the other varaible can be any & all values along the constant's axis. Thus, x & y are independent of each other.

#### Linear Equation in Two Variables
* Standard Form: ax + by + c = 0, where a,b,c are real numbers and a,b are not both 0 
 * x and y are the "two variables", which mean there are infinitely many solutions. & their values are **dependent** on each other.
 * For every number that is substituted for x, there is a corresponding y value. 
* The graph of a linear equation in2V will always be a slanted line that passes through both axes. Any ordered pair solutions to the equation are points that lie along the same line. Thus, x & y are dependent on each other.

## Intercepts
* The points at which a line intersect the x- & y-axes.
 * The **x-intercept** is the point, (a,0), where the graph crosses the x-axis. The x-intercept occurs when y is zero.
 * The **y-intercept** is the point, (0,b), where the graph crosses the y-axis. The y-intercept occurs when x is zero.
* Can graph a line using the intercepts from its equation.

#### Calculating Intercepts with Equation
* We can find the x & y intercepts by using a line's equation & this is often the easiest way to graph a linear equation.
* Since the x-intercept occurs when y is zero and the y-intercept occurs when x is zero gives, to find the x-intercept, let y=0 and solve for x. To find the y-intercept, let x=0 and solve for y.
* Example: Find the intercepts of 2x + y = 6
 * Y-Intercept:
   * x = 0
   * 2(0) + y = 6
   * y = 6
   * (0,6)
 * X-Intercept: 
   * y = 0
   * 2x + 0 = 6
   * x = 3
   * (3,0)


## Slope
* Visualy, slope is the steepness of a line.
* Conceptually, slope is the ratio of vertical change to horizontal change between two points on the plane or a line. So, slope equals the ratio of the rise over the run, measuring the rate of change for a line.
* **Rise**: Vertical change of a line (y-change).
* **Run**: Horizontal change of a line (x-change).
* **m**: Variable that's always used to represent slope.
* Formula: $m = \frac{rise}{run}$

* A line can slant upwards or downwards, but it is always read from left to right. Therefore, the rise can be positive or negative, but the run will always be positive. Thus, the rise determines the sign of a slope.
 * If a line slants upward, the slope is positive. 
 * If a line slants downward, the slope is negative.
* Slope is the hypotenuse of a right triangle, where the height & length of the sides are the rise & run, respectively. On a grid, if a line is vertical or horizontal, a right triangle cannot be formed because either the rise (height) or the run (length) is 0.
 * Horizontal lines have a slope of 0 because the rise is 0 and $\frac{0}{x} = 0$ (0 divided by anything is 0).
 * Vertical lines have a slope of 'undefined' because the run is 0 and $\frac{y}{0} = undefined$ (you can't divide by 0).

#### Determining Slope/Points/Lines Visually
* The slope of a line on a grid can be determined by counting the rise & run from one point to the other.
 * Count the Rise: Start from the leftmost point and count the intervals vertically to the second point.
  * If you count upward, the rise is positive.
  * If you count downward, the ries is negative.
 * Count the Run: From the point you're at, count the intervals horizontally to the second point.
 * Put the rise over run in a fraction.
* Likewise, a point on grid can be determined by counting in the same way using slope from another point. 
 * For graphs with larger numbers, it's helpful to use a larger equivalent fraction of the slope for accuracy.

#### Slope Formula
* If you have the coordinates of two points, you can get the slope by subtracting the first point from the second point to get the rate of change. 
* Formula: $m = \frac{y_2 - y_1}{x_2 - x_1}$
* You will get the same answer regardless of which point is subtracted from which.
* Example: Find the slope of (1, 2) and (3, 5)
 * $m = \frac{5-2}{3-1}$
 * $m = \frac{3}{2}$
* The slope formula is the same as $m = \frac{rise}{run}$ because rise represents a 'change in y' and run represents a 'change in x'.

#### Calculating Points Given a Point & Slope
* If you have the coordinates of 1 point and the slope of a line, you can calculate other points on the line to graph, since y is associated with the rise & x with the run.
* Unofficial Formula: $(x_1,y_1) = (x + run, y + rise)$
* To get the second point for a graph, you add the slope to the given point by adding the rise to y and the run to x. You can add the slope again to the result for a third point.
* Example: For a point, (3, 2), and a slope of 2, find another point.
 * $(x_1,y_1) = (3 + 1, 2 + 2)$
 * $(x_1,y_1) = (4,4)$
 * $(x_2,y_2) = (4 + 1, 4 + 2)$
 * $(x_2,y_2) = (5,6)$
 
#### Slope-Intercept Form
* Used to easily determine slope & y-intercept.
* The standard form of a linear equation in two variables is ax + by + c = 0, where x & y are the variables. But a linear equation in two variables can be written in different forms as long as there are two variables, ex: y = ax + b.
* Slope-intercept form, or "y-form", is a linear equation in two variables that's formatted as y = mx + b
 * When a linear equation is solved for y, the coefficient of the x term is the slope, represented as m rather than 'a', and the constant term, b, is the y-coordinate of the y-intercept. 
* To convert an equation to slope-intercept form, simply solve for y.
* Any equation formatted in this way automatically gives you a point & the slope.
* Example: For the equation $y = \frac{1}{2} + 3$
 * $m = \frac{1}{2}$
 * y-intercept = (0,3)

#### Point-Slope Formula
* Used to convert points into an equation.
* If you have the coordinates of 1 point and the slope of a line, you can calculate the equation of a line.
* Formula: $y - y_1 = m(x - x_1)$
* Example: Find the equation for a line with a point (3, 2) and a slope of 2.
 * y - 2 = 2(x - 3)
 * y - 2 = 2x - 6
 * y = 2x - 6 + 2
 * y = 2x - 4
* Since the equation is in slope-intercept form, we have another point, the y-intercept (0,4), to graph a line.

#### Parallel Lines
* Two lines that never intersect & have the same slope. 
* Denoted by comparisons:
 * is parallel with: $\|$
 * is NOT paralell with: $\nparallel$
 * the slope parallel to another line: $m_\|$
* Two lines that are linear equations in two variables are parallel when they have the same slope with different y-intercepts.
 * You can determine if lines are paralell by putting them into slope-intercept form. The slope should be the same, but the y-intercepts should be different.
* Two lines that are linear equations in one variable, using the same variable, are always parallel.
 * Horizontal lines, e.g. y = 3 and y = -6, always have a slope of 0.
 * Vertical lines, e.g. x = -2 and x = 9, always have a slope of undefined.

#### Perpendicular Lines
* Two lines that intersect at a 90$^{\circ}$ angle, i.e. their slopes are negative reciprocals of each other. 
* Denoted by comparisons:
 * is perpendicular to: $\perp$
 * is NOT perpendicular to: $\not\perp$
 * the slope perpendicular to another line: $m_\perp$
* Two lines that are linear equations in two variables are perpendicular when $m_1 * m_2 = -1$ and $m_1 = \frac{-1}{m_2}$.
 * Determine if lines are perpendicular if one slope is the reciprocal of the other & they have opposite signs. If you multiply their slopes, the result is -1.
 * Example: $y = \frac{1}{4}x - 1$ and $y = -4x + 2$
* Two lines that are linear equations in one variable, with different variables, are always perpendicular.
 * x = 3 and y = 5 are perpendicular.

# Linear Inequalities
* Linear inequalities are very similar to linear equations.
 * The standard forms are the same except with inequality relationship symbols rather than an equal sign.
* To visualize linear inequalities, we can use previous concepts. The point x=3 on a number line separates the number line into two parts, all values greater than 3 vs all values less than 3. Similarly, a line y=x+4 separates a plane into two regions.
* A **boundary line** is a line with the equation Ax+By=C that separates the region where Ax+By > C from the region where Ax+By < C.
* Unlike regular linear equations where all the solutions are on the same line, linear inequalities will have all solutions on one side of the line, and possibly on the line depending on if the inequality includes an 'or equal to' symbol or not.
 
 
## Solving Linear Inequalities in Two Variables
* Unlike regular linear equations in2V, where we choose any number for x & solve for y, linear inequalities in2V can be solved by choosing any values for x and y and evaluating to see if it's a true statement. Since there are infinite possibilites for x and y from which we can choose, best methods are to either use slope-intercept form and deduce logically, or to graph the boundary line and use test points from either side of the line. 

#### Slope-Intercept Method for Inequalities
* Put the inequality into slope-intercept form (y = mx + b), replacing equals sign with the inequality. Make sure to adhere to the rules for solving inequalities (flip the sign when multiplying/dividing by a negative).
   * The y-coordinate is whatever the inequality sign is to the y-intecept, b.
   * The x-coordinate can be 0.

#### Test Points Method
* You can choose any values for x & y to see if the statement is true, but if we graph the boundary line, we can choose points on either side of the line to evaluate and test if the inequality is true.
* Graph the boundary line (as if the inequality were an equality, i.e using the methods for linear equations).
* Choose a point to test, usually (0,0) as it's the easiest to evaluate.
 * Example: $x-2y<5$
   * Plug (0,0): $0-2(0)<5$
   * Evaluate: $0<5$, which is a true statement
   * Solution: (0,0) is a solution.
* If the x,y values you choose evalute to an equality, i.e. both sides of the equation are the same number, then the point is on the line & you'll have to choose a different point if the inequality is exclusive.
 * If (0,0) is on the line, i.e. the line passes through the origin, choose a point from the graph that isn't on the line. Without a graph, keep the y-coordinate & use oposites of the x-axis until a solution is found.
 * Example: $y \le -4x$
   * Plug (0,0) & Evaluate: $0 \le -4(0)$ and since 0 = 0, we need to choose another point.
   * Plug (1,0): $0 \le -4(1)$ is a false statement, choose another point.
   * Plug (-1,0): $0 \le -4(-1)$ is a true statement, thus is a solution.

# System of Equations
* When 2 or more linear equations are grouped together, they form a system of linear equations, denoted with a brace expanding to all equations.
* System of equations can be solved if at least as many unique equations exist as variables in the system of equations.
* Looking to find the values of the variables that are solutions to **all** equations and are represented by an ordered pair.

## Classifying Systems of Equations in Two Vars
* A system of equations with at least 1 solution is called a **consistent system of equations**, which includes intersecting lines and conincident lines.
* A system with no solutions is called an **inconsistent sytem of equations**, which are parallel lines.
* Also categorize the equations in a system of equations by calling the equations independent or dependent. 
* If two equations are **independent**, they each have their own set of solutions, such as intersecting lines and parallel lines.
* If two equations are **dependent**, all the solutions of one equation are also solutions of the other equation, such as coincident lines.

#### Inferring Classification
* If the slopes are different, the lines intersect. They are therefore consistent & independent.
* If the slopes are the same, but the y-intercept is different, the lines are parallel and are therefore inconsistent & independent.


## Graphs of System of Equations in Two Vars
* For a system of two equations, graph two lines. Can then see all the points that are solutions to each equation. The points that the lines have in common are solutions to the system.
* 3 possible cases:
 * Intersecting lines: 1 solution at the intersecting point.
 * Coincident lines: lines with the same slope & y-intercept are the same line & will have infinite solutions.
 * Parallel lines: never intersect & so have no solution.


## Solving via Substitution
* Solve for one variable first, plug it in to the other equation & solve for the other var, then plug back into 1st equation.
* Choose which variable to solve first in a way that will keep work easy.
* Use this method when one equation is already solved or can easily be solved for one variable.
 * **Example:** Solve for **x** & **y** in the equations **2x + 2y = 14** and **x - y = 3**. <br>

> EQUATIONS: <br>
$\begin{equation*}
2x + 2y = 14 \quad\text{(Equation 1)} \\
x - y = 3 \quad\text{(Equation 2)} 
\end{equation*}$

> PROCESS: <br>
$\begin{align*}
x - y &= 3 \quad\text{(Equation 2: solve for x)} \\
x &= y + 3 \quad\text{(add y to both sides)} \\
\\
2x + 2y &= 14 \quad\text{(Equation 1: substitue x for E2 value & solve for y)} \\
2(y+3) + 2y &= 14 \quad\text{(replace x with y+3)} \\
2y + 6 + 2y &= 14 \quad\text{(simplify with distribution)} \\
4y + 6 &= 14 \quad\text{(add like terms)} \\
4y &= 8 \quad\text{(minus 6 from both sides)} \\
y &= 2 \quad\text{divide by 4 on both sides} \\
\\
x &= y + 3 \quad\text{(Equation 2: substitue y for E1 value & solve for x)} \\
x &= (2) + 3 \quad\text{(plug in 2 for y)} \\
x &= 5 \quad\text{(simplify)}
\end{align*}$

> SOLUTION: (5,2)


## Solving via Elimination
* Uses the Addition Property of Equality to add the two equations together in a way that would eliminate one variable, i.e. have the coefficients of one variable be opposites so that they're cancelled out when added together.
 * Addition Property of Equality: For any expression a,b,c,d, if a=b and c=d, then a+c=b+d.
* Put both equations in standard form, then can apply other properties of equality to one or both equations to make a chosen variable's coefficients opposites. Add the equations together, solve for the remaining variable, and substitute that back in to the original equation.
* Use this method when the equations are in standard form.
 * Example: Solve 2x + y = 7 and x - 2y = 6. <br>

> EQUATIONS: <br>
$\begin{equation*}
2x + y = 7 \quad\text{(Equation 1)} \\
x - 2y = 6 \quad\text{(Equation 2)} 
\end{equation*}$

> PROCESS: <br>
$\begin{align*}
2(2x + y) &= (2)7 \quad\text{(Equation 1: multiply both sides by 2 to get y's opposite)} \\
4x + 2y &= 14 \quad\text{(Equation 1: simplify)} \\
\\
4x + 2y &= 14 \quad\text{(add both equations together)} \\
x - 2y &= 6 \\
---- &= --\\
5x ..... &= 20 \quad\text{(y is cancelled out)} \\
\\
5x &= 20 \quad\text{(simplify for x)} \\
x &= 4 \\
\\
(4) - 2y &= 6 \quad\text{(plug x back into one of the equations to get y)} \\
y &= -1 \quad\text{(simplify for y)}
\end{align*}$

> SOLUTION: (4,-1)

* When the system contains fractions, we first use the method to clear the equation of fractions (see Algebra-1 notebook) by multiplying each equation by the LCD of the fractions.
 * Example: Solve $x+\frac{1}{2}y = 6$ and $\frac{3}{2}x + \frac{2}{3}y = \frac{17}{2}$. <br>

> EQUATIONS: <br>
$\begin{equation*}
x+\frac{1}{2}y = 6 \quad\text{(Equation 1)} \\
\frac{3}{2}x + \frac{2}{3}y = \frac{17}{2} \quad\text{(Equation 2)} 
\end{equation*}$

> PROCESS: <br>
$\begin{align*}
2(x+\frac{1}{2}y) &= 2(6) \quad\text{(Equation 1: clear fraction by multiplying by LCD, 2)} \\
2x + y &= 12 \quad\text{(Equation 1: simplify)} \\
\\
6(\frac{3}{2}x + \frac{2}{3}y) &= 6(\frac{17}{2}) \quad\text{(Equation 2: clear fraction by multiplying by LCD 6)} \\
9x + 4y &= 51 \quad\text{(Equation 2: simplify)} \\
\\
-4(2x + y) &= -4(12) \quad\text{(Equation 1: multiply both sides by -4 to get y's opposite)} \\
-8x - 4y &= -48 \quad\text{(Equation 1: simplify)} \\
\\
-8x - 4y &= -48 \quad\text{(add both equations together)} \\
9x + 4y &= 51 \\
----- &= ---\\
x ..... &= 3 \quad\text{(y is cancelled out)} \\
\\
2(3) + y &= 12 \quad\text{(plug x back into one of the equations to get y)} \\
y &= 6 \quad\text{(simplify for y)}
\end{align*}$

> SOLUTION: (3,6)

# Evaluating Polynomials
* This section covers evaluation of binomials, trinomials, and higher. For evaluation of monomials, see Pre-Algebra notebook.
* Non-monomial polynomials are expressions whose terms are separated by a plus or minus sign.


## Adding/Subtracting Polynomials
* The polynomial will be wrapped in parentheses separated by an plus or minus operation.
* The Commutative Property allows us to rearrange the terms to put like terms together, but the operation sign must be distributed into the polynomial first (see distribution below for details).
 * Note that sign distribution is only necessary for subtracting polynomials, since multiplying a positive will not change the sign.
* Thus, adding/subtracting polynomials is just adding and subtracting a series of monomials. 
 * Example: $(9w^2−7w+5)−(2w^2−4) = 7w^2−7w+9$
   * Note that the -4 became positive after distributing the minus sign.
   
   
## Multiplying Polynomials
* We use distribution to get rid of parenthesis surrounding binomials when simplifying expressions.
* Algebraic multiplication on a binomial can occur using the Distributive Property with monomials or other binomial, trinomial, or polynomial expressions. 

### Multiply Monomials with Polynomials 
* Multiplying each of the terms in the polynomial within the parentheses by another term (monomial) outside the parentheses is called **Distribution**.
 * Example: 3(x + 2) = 3x + 6
 * Example: 4(x - 8)-(x + 3) = 4x - 32 - x - 3 = 3x - 35
 * Example: $3x^3y(x^2 - 8xy + y^2) = 3x^5y - 24x^4y^2 + 3x^3y^3$

### Multiply Binomials with Binomials
* Multiplying by another expression (binomial) in parenthesis.
* **FOIL** (the acronym for first, outer, inner and last) method is an efficient way of remembering how to multiply two binomials in a very organized manner. In FOIL, "First" means to multiply the terms which occur in the first position of each binomial. "Outer" means to multiply the first term in the first binomial with the last term in the last binomial. Etc.
 * Example: (2x + 9)(3x + 4)
   * First: 2x * 3x, Outer: 2x * 4, Inner: 9 * 3x, Last: 9 * 4
   * Evaluate: $6x^2 + 8x + 27x + 36$
   * Simplify: $6x^2 + 35x + 36$
* Other methods can be used for binomials with **special products**. Types of special products are 'square of sum', 'square of difference', and 'a difference of squares'.

#### Squaring Binomials
* Squaring a binomial multiplies it by itself and is called a **binomial square**, which produces a **perfect square trinomial** (discussed further in Factoring Expressions section below).
 * For a binomial square with a positive sign, the special product is called **square of sum**: $(a + b)^2 = a^2 + 2ab + b^2$
 * For a binomial square with a negative sign, the special product is called **square of difference**: $(a - b)^2 = a^2 - 2ab + b^2$
* Can solve using foil, but if you recognize the pattern, the shorter method is to sqare the first binomial term for the first trinomial term, sqare the second binomial term for the last trinomial term, and multiply the first & second binomail term & double it for the middle trinomial term. Put it in standard form.
 * Example: $(x + 9)^2$
   * Square 1st term: $x^2$
   * Square last term: 81
   * Multiply both & double: 2(9x) = 18x
   * Plug results: $x^2 + 18x + 81$

#### Product of Conjugates
* Two binomials that have the same first term and the same last term, but one is a sum and one is a difference is called a **conjugate pair** and produces a special product called **difference of squares**: $(x - y)(x + y) = x^2 - y^2$
* Notice that the product of conjugate pairs doesn't result in a trinomial like the others because the middle terms cancel each other out, resulting in a difference of two squares (discussed further in Factoring Expressions section below).
* Evaluate these simply by sqaring the first & last terms, separating them by a minus sign.
 * Example: $(2x - 3y)(2x + 3y) = 4x^2 - 9y^2$
   
### Multiplying Polynomials with Polynomials 
* Multiply each term with each other term.
* The FOIL Method only works binomial to binomial, but the vertical method or distribution works for everything.

#### Distributive Property Method
* Split up a polynomial into monomials that are distributed each into the other polynomial.
* Example: $(b+3)(2b^2-5b+8)$
 * Distribute 1st term: $b(2b^2-5b+8) = 2b^3 - 5b^2 + 8b$
 * Distribute 2nd term: $3(2b^2-5b+8) = 6b^2 -15b + 24$
 * Combine like terms: $2b^3 + b^2 - 7b + 24$
 
#### Vertical Method
* Use long multiplication method but for each term. Format one polynomial over the other & multiply down, lining up the partial products. For ease, put polynomial with fewer terms on bottom for less partial products.
* Example same as above: $(b+3)(2b^2-5b+8)$ <br>
$2b^2 -5b +8$ <br>
X .........$b +3$ <br>
.- - - - - - - - - - - - - -. <br>
+........$6b^2 -15b +24$ <br>
$2b^3 -5b^2 +8b$ <br>
.- - - - - - - - - - - - - -. <br>
$2b^3 +b^2 -7b +24$



## Dividing Polynomials
* Note that extended methods and advanced concepts are in 'Rational Expressions' section in Algebra-2 notebook.

### Dividing Polynomials by Monomials
* The method for dividing a polynomial by a monomial is based on fraction addition, where given 2 like fractions, the numerators are combined, but the denominators remain the same. The technique reverses this fraction addition property by splitting a fraction into two. I.e. To divide a polynomial by a monomial, divide each term of the polynomial by the monomial.
 * Example: $\frac{8y^2 + 24}{4} = \frac{8y^2}{4} + \frac{24}{4} = 2y^2 + 6$
 * Addition Exmaple: $\frac{x}{3} + \frac{2}{3} = \frac{x + 2}{3}$

* Multiplication Example: $\frac{12}{5}(-20x) = \frac{12}{5}(-\frac{20x}{1}) = -\frac{240x}{5} = -48x$

### Dividing Polynomials by Binomials
* The method of division will depend on either the sequence of degrees or divisor expression form. You can use either long division method or synthetic division method.
* Always put the expressions in standard form before dividing.

#### Long Division with Sequential Degrees
* If when in standard form, the dividend has descending degrees of 1, there will be no remainder.
* Example: $(x^2+9x+20)\div(x+5)$ <br>
 * Put in long div format: <br>
$x+5\overline{) x^2+9x+20}$ <br> <br>
 * Divide only the 1st divisor term, x, by the 1st dividend term, x$^2$, put quotient on top, & multiply that quotient (x), by the entire divisor expression back in. Subtract partial quotient from dividend & repeat with 1st divisor term again.<br>
.......................$x + 4$ <br>
$x+5\overline{) x^2+9x+20}$ <br>
.......$-(x^2 + 5x)$ <br>
.- - - - - - - - - - - - - - - -. <br>
.....................$4x +20$ <br>
.................$-(4x + 20)$ <br> 
.- - - - - - - - - - - - - - - -. <br>
....................= 0, no remainder

#### Long Division with Missing Degrees
* Dividend polynomials with missing sequences of degrees require place holders with a 0 coeffiecent for any of the missing degrees. Their remainder is written as a fraction with the divisor in the denominator.
* Example: $(x^4-x^2+5x-6)\div(x+2)$ is missing an $x^3$ term. <br>
 * Put in long div format, with the missing term place holder: <br>
$x+2\overline{) x^4+0x^3-x^2+5x-6}$ <br> <br>
 * Divide only the 1st divisor term, x, by the 1st dividend term, x$^4$, put quotient on top, & multiply that quotient ($x^3$), by the entire divisor expression back in. Subtract partial quotient from dividend & repeat with 1st divisor term again.<br>
.......................$x^3 -2x^2 +3x -1 -\frac{4}{x+2}$ <br>
$x+2\overline{) x^4+0x^3-x^2+5x-6}$ <br>
.......$-(x^4 + 2x^3)$ <br>
.- - - - - - - - - - - - - - - - - - - - - - - -. <br>
................$-2x^3 -x^2$ <br>
..............$-(2x^3 -4x^2)$ <br> 
.- - - - - - - - - - - - - - - - - - - - - - - -. <br>
...............................$3x^2 +5x$ <br>
...........................$-(3x^2 +6x)$ <br> 
.- - - - - - - - - - - - - - - - - - - - - - - -. <br>
.......................................$-x -6$ <br>
...................................$-(-x -2)$ <br> 
.- - - - - - - - - - - - - - - - - - - - - - - -. <br>
.............................................= $-4$ remainder, quotient written as a fraction with the divisor as the denominator.

#### Synthetic Division
* Used only for expressions where the divisor is of form x - c.
* Uses patterns to remove tedious steps & repeated characters of long division.
* Start with all the coefficients/constants from the dividend. Use the opposite sign of the constant from the divisor to multiply sums of products. Instead of dropping products down, move them to the side to be added to the next term's coefficient. Multiplying against the bottom row sums & putting products in the middle. The last sum is the remainder and the sums are the coefficients once the variables/exponents are added back from the right.
* Just as above, use 0 as a place holder for missing degree sequences.
* Example: $x^4 −16x^2 +3x +12 \div x + 4$
 * Take the opposite divisor coefficient: $-4$
 * Take  coefficients from the dividend, adding in a zero for the missing sequential degree: $1, 0, -16, 3, 12$
 * Drop down the first coefficient to the bottom row, multiply it by the divisor, and place in the second row to be added to the 2nd coefficient. Drop down the sum & multiply with the divisor again. Repeat. <br>
 
 $-4 | 1, -16, 3, 12$
 
| -4* | 1 | 0 | -16 | 3 | 12 |
|:---|---:|---:|---:|---:|---:|
| + |  | -4 | 16 | 0 | -12 |
| --- | --- | --- | --- | --- |
|  | 1 | -4 | 0 | 3 | 0 |

 * The last carry-down value is the remainder: 0
 * Put variables back from the right, exponents ascending (skipping the remainder): $+x^3 -4x^2 +3$
   * Note that the 0 coefficient will have exponent of 1, $0^1$, which is 0.


## Factoring Polynomials 
* Factoring expressions is useful in many situations including isolating/manipulating a variable in a polynomial that occurs in multiple unlike terms (like $x^2 + 3x$), or for simplifying polynomial fractions (covered in Rational Expressions section in Algebra-2 notebook), and more.
 * Although factoring a monomial expands it, factoring a polynomial usually shortens the polynomial.
* Factoring can result in a monomial GCF multiplied by an equation, a binomial multipied by a binomial, or both.
 * Sometimes you may need to factor twice if the resulting expression can be factored again.
* Not all polynomials can be factored. Just like some numbers are prime, so too are some polynomials prime. 
* A polynomial is completely factored if, other than monomials, its factors are prime.
* Use ditribution or foil to check that you get the original equation after solving.
* There are many methods for factoring polynomials.
 * Use the strategy list at the bottom of this section to determine the best factoring method to use.

### Factor Polynomials With GCF
* If all terms in the polynomial share a GCF, use the Distributive Property in reverse. 
* Always check for GCF first before and factor it out before moving on to other methods.
* The GCF can be a polynomial.

 1. Find the GCF between all terms in the expression (see Pre-Algebra notebook).
   1. If the leading coefficient is negative, turn the GCF negative.
 2. Divide each term by the GCF.
 3. Put the divided terms in parenthesis with the GCF on the outside.

> Example: Factor $3x^2 + 6x + 9$
 * GCF: 3
 * Divide each by 3: $x^2, 2x, 3$
 * Multiply by GCF: $3(x^2 + 2x + 3)$
 
### Factor Quadrinomials+
* Polynomials with 4 terms that do not share a GCF may have resulting binomial factors or trinomial factors even.
* For 4+ terms use **Factor By Grouping**, where you group terms to find the factor between those.
* May have to try different combinations of grouping and/or group negative signs in a way that would match factors so that they can be distributed in reverse. 
* Usually group into pairs of terms as exampled below, but can also group a trinomial and use trinomial method to solve, which will be shown in the mixed methods section at the end of this section.
* Example: xy + 3y + 2x + 6
 * Group: (xy + 3y) + (2x + 6)
 * Factor the GCF from each pair: y(x + 3) + 2(x + 3)
 * Rewrite using distribution: (x + 3)(y + 2)
* Example: $18x^2 -27xy -10xy +15y^2$
 * Group 1st sets: $(18x^2 -27xy) + (-10xy +15y^2)$
 * Factor GCF from each pair: $3x(6x -9y) + 5y(-2x +3y)$ does not yeild a common factor, try another grouping.
 * Group swapped middle: $(18x^2 -10xy) + (-27xy +15y^2)$
 * Factor GCF from each pair: $2x(9x -5y) + 3y(-9x +5y)$ does not yeild a common factor because of the negative.
 * Factor considering negatives: $2x(9x -5y) + (-3y)(9x -5y)$
 * Rewrite using distribution: $(2x -3y)(9x -5y)$
   * Note that this could have also been solved by combining like terms and solving as trinomial with methods below, but not all quadrinomials will.
 
### Factor Quadratic Trinomials
* Polynomials with three terms and a degree of 2 will have resulting binomial factors instead of a monomial to polynomial expression of factors.
* **Sign rules**: 
 * If only the middle term is negative (last term positive), the binomials both have minus signs.
   * Example: $t^2 - 11t + 28 = (t - 7)(t - 4)$
 * If the last term is negative (regardless of the middle), the binomials have opposite signs. To determine which has which sign, match the sign of the middle term with the number of the highest absolute value.
   * Example: $x^2 + 4z - 5 = (z - 1)(z + 5)$
   * Example: $x^2 - 4z - 5 = (z + 1)(z - 5)$
* Different methods based on the form of the expression:
 * If the trinomial is a perfect square trinomial, use the patterns of special products.
 * If the trinomial is quadratic with a leading coefficient of 1, use the methods according to specific forms.
 * If the trinomial has a leading coefficient > 1, use the tedious trial and error method by undoing FOIL.

#### Factor Trinomials with Leading Coefficient 1
* There are two quadratic trinomial forms for which this method can be applied. 
* **Form 1**: $x^2 + bx + c$, is factored by starting with two empty parens, () (), and factoring each term into those. The 1st binomial terms will always be the same because it's quadratic (squared). The last terms in the resulting binomials will be the factors of the last trinomial term where their sum equals the middle trinomial term (the middle term comes from adding the outer & inner binomial terms during FOIL). 
 * Example: Product = $x^2 + 5x + 6$
   * Factor 1st term in equation: (x ) (x )
   * List factors of last term in equation: 1 * 6, 2 * 3
   * Use factors that sum to middle term: 2 + 3 = 5
   * Factors = (x + 2)(x + 3)
* **Form 2**: $x^2 + byx + cy^2$, factored the same way as form 1 except the last binomial terms will include y.
 * Example: Product = $x^2 + 12xy + 36y^2$
   * Factor 1st & last(variable) terms: (x + y) (x + y)
   * List factors of last term coefficient: 36 * 1, 2 * 18, 6 * 6, 3 * 12
   * Use factors that sum to the middle term: 6 + 6 = 12
   * Factors = (x + 6y)(x + 6y)

#### Factor Perfect Square Trinomials
* Types of **special products** that produce a **binomail squares** which can either be a **square of a sum** or a **square of a difference**. Perfect Square Trinomial can be factored using the methods above (quadratic, leading co. 1), but spotting the pattern can save time instead. 
* **Square of a Sum Form**: $a^2 + 2ab + b^2 = (a + b)^2$
* **Square of a Difference Form**: : $a^2 - 2ab + b^2 = (a - b)^2$
 * Note the middle term is negative, thus the binomail result is a difference squared.
* To identify this pattern, check that the 1st and last terms are square roots. Multiply those roots together and by 2 to equal the middle.
* To factor, plug the square roots of the 1st and last terms into the binomial. The middle term sign of the original determines if the binomial is a difference(-) or sum(+). Then square it.
* Example: $y^2 + 14y + 49$
 * Form: $(y)^2 + 2(y)(7) + (7)^2$
 * Factor: $(y + 7)^2$
* Example: $9x^2 - 30xy + 25y^2$
 * Form: $(3x)^2 - 2(3x)(5y) + (5y)^2$
 * Factor: $(3x - 5y)^2$
 
#### Factor Trinomials with Leading Coefficient Greater Than 1
* There are 2 methods which can by applied. If the equation has higher coefficients, it may be quicker to use trial & error. Below are the same 2 examples solved using both methods. 
* **Form**: $ax^2 + bx + c$ is factored using trial & error if there is no GCF.
* Expect binomial factors and consider the signs of the terms. 
* **Trial & Error Method**: List all the factors of the 1st term and last terms. Swap them out in the resulting binomial and do foil until the result equals the original equation. To save time, we don't need to consider combinations where either binomial has a common factor in its terms because we already know there is no GCF.
 * Example: $18x^2 −37xy +15y^2$
   * Consider signs: (-)(-) since middle term is negative.
   * List factors of 1st term: 18x * x, 2x * 9x, 3x * 6x
   * List factors of last term: 15y * y, 3y * 5y
   * Attempt 1: $(18x-15y)(x-y)$ is false because 18 and 15 share a factor of 3.
   * Attempt 2: $(18x-y)(x-15y) = 18x^2 -271xy +15y^2$ is false.
   * Attempt 3: $(18x-3y)(x-5y)$ is false because 18 and 3 share a factor of 3.
   * Attempt 4: $(18x-5y)(x-3y) = 18x^2 -59xy +15y^2$ is false.
   * Attempt 5: $(2x-y)(9x-15y)$ is false because 9 and 15 share a factor of 3.
   * Attempt 6: $(2x-15y)(9x-y) = 18x^2 -137xy +15y^2$ is false.
   * Attempt 7: $(2x-3y)(9x-5y) = 18x^2 -37xy +15y^2$ is true.
 * Example: $6x^2 +7x +2$
   * Consider signs: (+)(+)
   * List factors of 1st term: x * 6x, 2x * 3x
   * List factors of last term: 1 * 2
   * Attempt 1: $(x + 1)(6x + 2)$ is false because 6 and 2 share a factor of 2.
   * Attempt 2: $(x + 2)(6x + 1) = 6x^2 +7x +2$ is false.
   * Attempt 3: $(2x + 2)(3x + 1)$ is false because 2 and 2 share a factor of 2.
   * Attempt 4: $(2x + 1)(3x + 2) = 6x^2 +7x +2$ is true.
* **AC Method**: Multiply first term, a by last term, c. Find factors of that product that add to the middle term, b. Rewrite the middle term with those factors and the use the "factor by grouping" method.
 * Example: $6x^2 +7x +2$
   * Consider the signs: (+)(+) since both terms are positive.
   * Multiply first & last coefficients/constants: 6 * 2 = 12
   * List factors of product: 2 * 6, 3 * 4, 12 * 1
   * Find factors that sum to middle term: 3 + 4 = 7
   * Rewrite expression with factored middle: $6x^2 +4x + 3x +2$
   * Factor by grouping: $3x(2x +1) + 2(2x +1)$ (swapped middle terms since the other way cannot be factored)
   * Rewrite using common factor: $(2x +1)(3x +2)$
 * Example: $18x^2 −37xy +15y^2$
   * Consider the signs: (-)(-) since middle is negative.
   * Multiply first & last: 270
   * List factors of product: 1 * 170, 2 * 135, 3 * 90, 5 * 54, 6 * 45, 9 * 30, 10 * 27, 15 * 18
   * Find factors that sum to middle: 10 + 27 = 37
   * Rewrite expression with factored middle: $18x^2 -27xy -10xy +15y^2$  
   * Factor by grouping (see ex in that section): $18x^2 -10xy -27xy +15y^2$
   * Group: $(18x^2 -10xy) + (-27xy +15y^2)$
   * Factor each pair: $2x(9x -5y) + 3y(-9x +5y)$
   * Rewrite using distribution: $(2x -3y)(9x -5y)$
   
#### Transform Expressions by Factoring using Substitution
* Sometimes an expression does not appear to be in the $ax^2+bx+c$ form. However, when the second variable is still a square root of the 1st variable, as it is in the $ax^2+bx+c$ form ($x$ is a square root of $x^2$), we can use a method called **factor using substitution**. 
* Substitute the root and factor using previous methods for that form. It is standard to use the letter, u, for the substitution.  
* Example: $x^4 - 4x^2 -5$
 * Substitute the square root: $u = x^2$
 * Rewrite using substitution: $u^2 -4u -5$
 * Factor: $(u+1)(u-5)$
 * Plug in original value: $(x^2+1)(x^2-5)$
* Example: $(x−2)^2 +7(x−2) +12$
 * Substitute the square root: $u = x-2$
 * Rewrite using substitution: $u^2 +7u +12$
 * Factor: $(u+3)(u+4)$
 * Plug in original value: $((x-2)+3)((x-2)+4)$
 * Simplify: $(x+1)(x+2)$
   * Note the 2nd ex can also be factored by multiplying out/distributing then combining like terms.
 
### Factor Binomials

#### Difference of Squares
* Binomial where 2 squares are separated by a minus sign is a type of **special product** known as the Product of Conjugates (discussed in multiplying binomials section above) and results in a difference of squares.
* **Difference of Two Squares Form**: $x^2 - a^2 = (x + a)(x - a)$
* To identify this pattern, check that both terms in the binomial are squares and they're separated by a minus sign.
* To factor, plug in the square root of the first term in each binomial and the square of the second term in each binomial. One binomial will be a plus and the other a minus.
* Example: $25x^2 - 81$
 * Form: $(5x)^2 - (9)^2$
 * Factor: $(5x + 9)(5x - 9)$
* Example: $64y^2 - 1$
 * Form: $(8y)^2 - (1)^2$
 * Factor: $(8y+1)(8y-1)$
* The **sum of two squares**, of the form $x^2 + a^2$, is prime and **cannot be factored** with real numbers. 

#### Sum & Difference of Cubes
* These are a type of **special product** that can be factored and using a pattern.
 * Unlike the 'sum of squares', the sum of cubes can be factored.
* **Sum of Two Cubes Form**: $a^3+b^3=(a+b)(a^2−ab+b^2)$
* **Difference of Two Cubes Form**: $a^3−b^3=(a−b)(a^2+ab+b^2)$
* To identify, check for cubes in each term of a binomial.
* To factor, memorize the forms and plug in the square roots: In each case, the sign of the original binomial matches that of the 1st binomial factor. And the 1st sign of the trinomial is opposite. The trinomial factor cannot be factored.
* Example: $x^3 + 64$
 * Identify signs: + is the sum of cubes
 * Form: $(x)^3 + (4)^3$
 * Factor: $(x + 4)(x^2 - 4x + 16)$
* Example: $27a^3 -125b^3$
 * Identify signs: - is the difference of cubes
 * Form: $(3a)^3 - (5b)^3$
 * Factor: $(3a - 5b)(9a^2 + 15ab + 25b^2)$
* Can also work with expressions & will require simplifying.
 * Example: $(x+5)^3 - 64x^3$
   * Factor: $((x+5) - 4x)((x+5)^2 + (x+5)(4x) + (4x)^2)$
   * Evaluate: $(-3x +5)(x^2 + 10x + 25 + 4x^2 + 20x + 16x^2)$
   * Simplify: $(-3x + 5)(21x^2 + 30x + 25)$

### Prime Polynomials
* Only way to be certain a trinomial is prime is to list all the possibilities of trial and error to show that none of them work, starting with the GCF.
* Use a table or list to show that each possibility was attempted.
* For trinomials with coefficient 1, use a table or list showing all the last term factors and their sums to see if the sum matches the middle. For trinomials with coefficient greater than 1, use trial & error list or ac grouping list, etc.
* Example: $x^2 -9xy -12y^2$
 * List last term factors as sums:
   * -1 + 12 = 11
   * 1 + (-12) = -11
   * -2 + 6 = 4
   * 2 + (-6) = -4
   * -3 + 4 = 1
   * 3 + (-4) = -1
 * Determine if any sums match middle term: $-9 \neq$ any sums.

### Factor Using Multiple Methods
* Sometimes expressions can be factored many times over.
* A polynomial may have a GCF which can be factored out, but the result can still be factored further.
 * Example: $2x^2y + 16xy + 30y$
   * GCF: 2y
   * Factor GCF: 2y(x^2 + 8x + 15)
   * Factor trinomial ($ax^2 +bx +c$ form): 2y(x + 3)(x + 5)
  * Example: $4x^2 +8bx −4ax −8ab$
     * Factor GCF: $4(x^2 +2bx −ax −2ab)$
     * Factor by grouping: 4( x(x + 2b) + -a(x + 2b) )
     * Rewrite: 4(x - a)(x + 2b)
 * Example: $48x^4y^2 -243y^2$
   * Factor GCF: $3y^2(16yx^4 - 81y)$
   * Factor difference of squares: $3y^2(4x^2 - 9)(4x^2 + 9)$
   * Factor difference of squares again (1st binomial): $3y^2(2x -3)(2x +3)(4x^2 + 9)$
     * Note the last binomial is a sum of squares which cannot be factored.
 * Example: $6x^3y + 48y^4$
   * Factor GCF: $6y(x^3 + 8y^3)$
   * Factor sum of cubes: $6y(x + 2y)(x^2 - 2xy + 4y^2)$
* A quadrinomial can have a special groupings in 3 rather than 2 terms.
 * Example: $x^2 −6x +9 −y^2$
   * Group trinomial (perfect square trinomial): $(x^2 -6x +9) -y^2$
   * Factor perfect square trinomial: $(x - 3)^2 -y^2$
   * Factor difference of squares: $((x-3) +y)((x-3) -y)$
   * Rewrite solution with variables first: $(x+y-3)(x-y-3)$

# Quadratic Equations
* A polynomial equation of degree two is called a quadratic equation.
 * Note as opposed to linear equations, which have only 1 degree.
* Of the form $ax^2 + bx + c = 0$, where a,b,c are real numbers and $a\neq0$.
 * Examples:
   * $x^2 + 5x + 6 = 0$
   * $3y^2 + 4y = 10$
   * $n(n + 1) = 42$
* Many methods for solving. Others discussed in Algebra-2 notebook.
 
## Solve Quadratics by Factoring
* To solve quadratic equations, you cannot just isolate the variable as done in linear equations. We can solve using the logic of factoring along with the zero product property.
* Recall Zero Product Property of real numbers: If ab = 0, then either a = 0 or b = 0 or both.
* Since multiplying any number by 0 results in 0, and a factored quadratic equation will have a product, if the product is 0, we know that one or both of the factors is 0. 
* To solve, put equation in standard form with zero isolated to one side. Factor the equation, then set each factor to 0 rendering 2 separate equations with the same variable. Solve for the variable in each for the solutions.
* Validate solutions by substituting each solution separately into the original equation to get 0.
 * Example: Solve $30n^2 -17n = -2$
   * Put in standard form: $30n^2 -17n +2 = 0$
   * Factor equation: $(5n-2)(6n-1)=0$
   * Rewrite 2 equations with factors: $5n-2=0$ and $6n-1=0$ Since the product is 0, one or both factors must be 0.
     * Solve for E1: $n=\frac{2}{5}$
     * Solve for E2: $n=\frac{1}{6}$
   * Validate E1: $(5(\frac{2}{5})-2)(6(\frac{2}{5})-1)=0$
   * Validate E2: $(5(\frac{1}{6})-2)(6(\frac{1}{6})-1)=0$
* If one side of the equation is already factored but the other side isn't zero, multiply the factors, set the equation to zero, and add like terms. Then factor again to find the solution.
 * Example: $(3x−8)(x−1) = 3x$
   * Multiply factors & set to 0: $3x^2 -3x -8x + 8 -3x = 0$
   * Combine like terms: $3x^2 -14x + 8 = 0$
   * Factor trinomial: $(3x−2)(x−4) = 0$
   * Continue to solve as above.
* Sometimes there may be 3 factors. If a factor is a constant, it cannot equal zero.
 * Example: $3x^2 −12x −63 = 0$
   * Factor GCF: $3(x^2 -4x -21)$
   * Factor trinomial: $3(x-7)(x+3) = 0$
   * Rewrite 3 equations: $3 \neq 0$ and $x-7 = 0$ and $x+3 = 0$
     * Solve for E2: $x = 7$
     * Solve for E3: $x = -3$

# Relations
* A **relation** is any set of ordered pairs, (x,y), that represents the relation between an input and an output, (input,output).
 * Set of all the input values (often x-values) in the ordered pairs together make up the **domain** of the relation. 
 * Set of all the output values (often y-values) in the ordered pairs together make up the **range** of the relation.
* Sets in general are denoted by wrapping in curly brackets.
* Relations can have duplicates, but the domain and range sets are written without duplicates, usually in ascending order.
 * Example: Find domain & range for the relation: {(1,5),(−3,−1),(4,−2),(0,3),(2,−2),(−3,4)}
   * Domain: {−3,0,1,2,4}
   * Range: {−2,−1,3,4,5}
 * Notice the domain/range sets only have 5 values, rather than 6 because repeat numbers were omitted. 
* Can also use interval notation to represent domain & range.
 * Example: Using the above relation:
   * Domain: [-3,4]
   * Range: [-2,5]
* A mapping is sometimes used to show a relation, whereby a table of mismatched domains & ranges is matched by arrows to show the pairing of the elements of the domain with the elements of the range.
* A graph can also be used to show relations by plotting the ordered pairs as points.


# Functions
* A function is a special type of relation that assigns to each element in its domain exactly one element in the range.
 * **input** ---> <u>|function|</u> ---> 1 **output**
* The same is not true for the range values, which can be assigned multiple domain values.
* In a function, for each x,y ordered pair in the relation, each x-value is matched with only one corresponding y-value so there are no repeated elements in the domain, but multiple domain values may exist for a single range value.
 * Example: A table of people matched with their birthdays is a relation, but it's also a function. The domain is their names & the range is their birthdays. A person only has one birthday, but multiple people can have the same birthday.
 * The example in the relations section above is not a function because there are repeated elements in the domain (set of x-values).
 
#### Function Notation
* We use function notation to illustrate this concept, making it easier to read & understand.
* It is very convenient to name a function and they're most commonly named f. Other common names are g, h, F, G, or H.
* The most common input name is x.
* For the function f, we write the range value y as $f(x)$, read "f of x" or "the value of f at x". So $y = f(x)$, read "y is a function of x".
 * Exmaple: The equation $y=4x-5$ where $y = f(x)$:
   * Written: $f(x)=4x-5$ since y is a function of x.
* The parentheses does not indicate multiplication; we're simply replacing the variable y with $f(x)$ to illustrate the functional relationship between the input and the output.
* We call x the **independent variable** as it can be any value in the domain. We call f(x) the **dependent variable** as its value depends on x.
 * This concept is seen with linear equations in2V where the value of y is dependent on x. You can write them using function notation, although it isn't necessary.
* The variables & functions can be named anything and represent many scenarios, as long as we're using parenthesis with a function to represent input and corresponding output.
 * Example: $h(a) =$ the next integer with the same 1st letter as input a.
   * $h(2) = 3$
   * $h(8) = 11$
* The graph of the function is the graph of all ordered pairs $(x,y)$ where $y=f(x)$, so we can write the ordered pairs as $(x,f(x))$.

#### Determining If An Equation Defines a Function
* To distinguish a function from a relation or determine if an equation represents a function, it's most helpful to use the slope-intercept form of the linear equation. The slope-intercept form of a linear equation is y=mx+b. In function notation, this linear function becomes $f(x)=mx+b$ where m is the slope of the line and b is the y-intercept.
* It is easiest to see if the equation is a function when it is solved for y. If each value of x results in only one value of y, then the equation defines a function.
 * Example: For 2x+y = 7, let x=3, which evaluates to y=1. Since y is one value, the equation is a function.
 * Example: For $x + y^2=3$, let x=2, simiplified is $y^2=1$, evaluates to $y = \pm\sqrt{1}$. Since square roots have 2 values, positive & negative, the equation is not a function of x.
 * Example: For $y=x^2+1$, let x=2, which evaluates to y=5. Since y is one value, the equation is a funciton.
* Thus, equations where y is squared or has an even power are not functions, but the same is not true for x to an even power since x can have multiple values for y.


## Evaluating Functions
* The process of finding the value of $f(x)$ (the output) for a given value of x (an input) is called **evaluating the function**.
* When substituting a value for x to determine the y-value, we replace all the x variables with the assigned value, including the x in the function name. Again, the name does not represent multiplication & does not affect the evaluation of the equation. It's only the name & the equation is evaluated per usual.
 * Example: For $f(x) = 4x-5$, evaluate $f(2)$
   * Written: $f(2) = 4(2)-5$ 
   * Evaluated: $f(2) = 3$
   * Read: "The value of the function at x=2 is 3."
* The same rules apply for substituting variables.
 * Example: For the function $g(x)=3x-5$, evaluate $g(h^2)$.
   * Evaluated: $g(h^2)=3h^2-5$
* Similarly, for expressions:
 * Example: For the function $g(x)=3x-5$, evaluate $g(x+2)$.
   * Evaluate: $g(x+2)=3(x+2)-5$
   * Simplify: $g(x+2)=3x+6-5$
   * Simplify: $g(x+2)=3x+1$
* When adding 2 functions as a function, evaluate both and add. Solve for any x constants first then add that to the original:
 * Example: For the function $g(x)=3x-5$, evaluate $g(x)+g(2)$.
   * Find $g(2)$ first: $g(2) = 3(2)-5$
   * $g(2) = 1$
   * Add: $g(x)+g(2) = 3x-5+1$, where the 1 at the end is g(2)
   * Simplify: $g(x)+g(2) = 3x-4$
     * Notice that $g(x+2) \neq g(x)+g(2)$
     

## Solving Functions for Inputs
* While evaluating functions generally means solving for $f(x)$ or y, solving for the inputs of a funtions means finding the value(s) of x when given the value for $f(x)$.
* To solve for the x inputs, replace the entire function name, $f(x)$, with the value given (rather than just replacing the x value when evaluating for y) and then solve for x. Like evaluating functions, finding x inputs for functions can be rewritten with the x value in the function name and the y value on the right hand side.

* Since functions will only have one y value for any x value, but can have many x values for any y value, many functions have higher degrees for x, i.e. quadratic+.
* Solving quadratic functions is similar to solving quadratic equations where you can use factoring and the zero product property after replacing the function name with the y-value. This will give you 2 values for x that correlate with 1 y value and can be used as ordered pairs for points on a graph.
* Example: Solve $f(x) = x^2 + 2x - 2$ when $f(x) = 6$ and identify 2 points on the graph of the function.
 * Replace function name: $6 = x^2 + 2x - 2$
 * Put in quadratic standard form (one side 0): $x^2 + 2x - 8 = 0$
 * Factor trinomial: $(x - 2)(x + 4) = 0$
 * Express factors as equations for 0: $x - 2 = 0$ and $x + 4 = 0$
 * Solve E1: $x = 2$
   * $f(2) = 6$
 * Solve E2: $x = -4$ 
   * $f(-4) = 6$
 * Ordered Pairs: 
   * $(2,6)$ and $(-4,6)$


## Graph of a Function
* Since a function can have multiple x-values to a y value but cannot have multiple y-values to an x-value, we can use the **vertical line test** to determine if a graph is a function. A vertical line drawn through the graph should only hit a y-value once.
* There will only be 1 y-intercept, but can have multiple x-intercepts.
  * x-intercepts occur when y=0, so the x-intercepts occur when $f(x)=0$.
  * y-intercept occurs when x=0, so the y-intercept occurs at $f(0)$
* The y-coordinate of a point is the function of its x-coordinate.
* The domain is between the leftmost & rightmost points, while the range is between the highest & lowest points. Each can be written using interval notation.

## Key Features of Functions

### Domain, Codomain, and Range
* **Domain**: All the possible inputs of a function as we define them.
* **Codomain**: All the possible outputs of a function as we define them.
* **Range**: Aka, image. The actual outputs of a function based on logic. The range is a subset of the codomain.

#### Importance of Codomain
* An equation may or may not be a function depending on the outputs we allow.
 * Example: The square root function may have a negative and positive output, but a function cannot have multiple outputs. In the case that only the square root symbol is used, this implies the principal (positive) square root so the codomain is correct, but in other cases, it may be necessary to **limit the codomain** to only certain types of numbers, e.g. positive or negative integers.
 
#### Notation
Specifying the domain & Codomain can be done in a number of ways, e.g. interval notation, set notation, etc.
* $f : \mathbb{N} \mapsto \mathbb{N}$ means a function, $f$, has the domain of natural numbers, and the codomain is of natural numbers also.
* {$x \in \mathbb{N} \mid x > 0$} means the set of all x is a member of the natural numbers such that x is greater than 0.
* $[0,\infty)$ means the lower bound is 0, inclusive, to infinity.


### Intercepts
* There will only be 1 y-intercept, but can have multiple x-intercepts.

#### x-intercepts
* x-intercepts occur when y=0, so the x-intercepts occur when $f(x)=0$, as the point (x,0) on a graph.
* The values of x where the function is 0, are also called the **zeros of the function** or the **roots**. 
* Found by setting the equation to 0 and solving for the x('s).
 * Example: For $f(x) = 3x^2 + 10x - 8$, find the zeros of the function.
   * Replace function name & reform: $3x^2 + 10x - 8 = 0$
   * Factor trinomial: $(3x-2)(x + 4) = 0$
   * Solve for x in E1 factor: $3x - 2 = 0$
   * Solve E1: $x = \frac{2}{3}$
     * $f(\frac{2}{3}) = 0$
   * Solve for x in E2 factor: $x - 4 = 0$
   * Solve E2: $x = 4$
     * $f(4) = 0$
   * Zeros of the function: $\frac{2}{3}$ and $4$
   * x-intercepts: $(\frac{2}{3}, 0),(4, 0)$

#### y-intercept
* y-intercept occurs when x=0, so the y-intercept occurs at $f(0)$
* The y-intercept is found by plugging in 0 for x.
 * Example: For $f(x) = 3x^2 + 10x - 8$, find the y intercept. 
   * Plug in 0: $f(0) = 3(0)^2 + 10(0) - 8$
   * Simplify: $f(0) = -8$
   * y-intercept: $(0,-8)$


### Extrema
* The maxima and minima (the respective plurals of maximum and minimum) of a function, known collectively as extrema (the plural of extremum), are the largest and smallest value of the function

* **Absolute extrema**: aka global extrema, max or min on the entire domain.
* Unbounded infinite sets, such as the set of real numbers, have no minimum or maximum.

### Discrete vs Continuous
* Some types of data can only be quantified using certain types of numbers, which affect the graph of the data.

#### Discrete
* **Discrete data** means the values in the set are unconnected, distinct entities & are countable. The type of data cannot be subdivided into different parts.
 * N, W, Z and Q (sets of Natural, Whole, Integer, and Rational Numbers) are referred to as discrete numbers because there are always other numbers in between them that separate them from each other. 
 * Rational numbers are discrete because they are separated by the irrational numbers creating infinitesimally small gaps between them, allowing them to be counted with natural numbers. But they are not used to describe discrete data unless they are fixed values.
 * Examples of discrete data are number of people, number of items, height of a building in stories, etc.
* **Discrete function**: a function where the domain is countable, having individual values or consisting of only certain numbers in an interval. The graph looks like a scatterplot (a bunch of points that are not connected together) and has no connecting line throughout.

#### Continuous
* **Continuous data** means the values in the data set can take on any value within a finite or infinite interval, cannot be counted, & is instead measured. The data can be subdivided into decimals or fractions.
 * R (set of Real Numbers) is continuous because between any two real numbers, however close they may be, there are infinitely more real numbers, thus there are never gaps and it's continuous. 
 * Examples of continuous data are time, temperature, height of a person, etc.
* **Continuous function**: a function where the domain is a range of all real numbers in an interval since every point has meaning to the original problem. The graph has a continuous line without gaps.

## Types of Function Mappings
* Functions can be characterized by the mappings of the relationships of their elements.
* Characterized by the uniqeness of the range values as well as whether they match the codomain.
* Note that the names of the types are not very intuitive. Although 'Into' means domain & range aren't equal, 'Onto' means the opposite as well as being a many-to-one.

![title](./../resources/func-mapping.png)

#### Many-to-One Functions
* The typical function definition, where domain values can only have 1 range value, but range values can have multiple domain values. 
* Many-to-One functions may or may not have matching range & codomains.
* In the picture above, the 'General Function' and 'Surjective' examples are Many-to-One functions.
* When there is only 1 element in the codomain, thus all elements in the domain are matched to it, the function is a constant function.
* Uses vertical line test reviewed above.

#### Into Function
* The range and the codomain are not equal. The range is a subset of the domain.
* Many-to-One function where there are leftover values in the codomain. 

#### Onto Functions (Surjective)
* There are no unmatched codomain values, i.e. the range and codomain are equal.
* A Many-to-One function but there are **not** leftover values in the codomain.

#### One-to-One Functions (Injective)
* The requirement of a function differentiating it from a relation is that each element in the domain can have only 1 corresponding range value. A one-to-one function extends that requirement to the range values as well.
 * A function is one-to-one if each value in the range corresponds to, at most, one element in the domain. For each ordered pair in the function, each y-value is matched with only one x-value and vice versa. There are no repeated y-values or x-values.
* There may also be unmatched codomain values, i.e. the range and codomain needn't be equal, and thus is also called an **injective** funciton.
* Just like the vertical line test on a graph can determine whether a relation is a function, the **horizontal line test** require a horizontal line to pass through only 1 y-value.
* Some example graphs of one-to-one functions can be viewed in the 'Types of Basic Functions'. Some examples are the linear, identity, cube, and square root functions.
 * Note that the parent functions pass, but not all of these types of functions may pass.

#### One-to-One Correspondence Function (Bijective)
* A function that is both 'one-to-one' and 'onto' function, i.e. both injective & surjective.
* Every element in the domain & codomain must have only 1 unique match, similare to a One-to-One.
* There are no unmatched codomain values, i.e. the range and codomain are equal.

## Composition of Functions
* An operation on functions where the output of one function is the input of a second function is called **composition**, denoted with a $\circ$ symbol, and creates composite functions.
* The composition of functions $f$ and $g$ where g is composed with f (or g's output is f's input), is written $f \circ g$ and is defined by $(f \circ g)(x) = f(g(x))$, read "f of g of x".
 * Note that the parenthesis around the function name are sometimes omitted.
* One example of composing functions is transformations (Quadratic Function section), whereby a basic graph function is shifted a number of units. For $g(x) = x^2$ and $f(x) = x - 4$, the composition is defined by $(f \circ g)(x) = f(g(x))$, giving an output of $f(g(x)) = x^2 - 4$.
![title](./../resources/composition-ex.png)


#### Evaluating Compositions 
* When evaluating composite functions, we're evaluating from the inside out of the parenthesis, or aka from right to left of the function definition. After evaluating the innermost function, it's output becomes the input of the function wrapping it. Thus the order of functions matters. 
 * Example: For $f(x) = 4x - 5$ and $g(x) = 2x + 3$, find $(f \circ g)(x)$ vs $(g \circ f)(x)$.
 * 1. $(f \circ g)(x)$ means $f(g(x))$ and $g(x)$ is the innermost known value.
     * Substitute $g(x)$: $(f \circ g)(x) = f(2x + 3)$
     * Plug g into f $g(x)$: $(f \circ g)(x) = 4(2x + 3) - 5$
     * Simplify: $(f \circ g)(x) = 8x + 7$
 * 2. $(g \circ f)(x)$
     * Substitute $f(x)$: $(g \circ f)(x) = g(4x - 5)$
     * Plug f into g: $(g \circ f)(x) = 2(4x - 5) + 3$
     * Simplify: $(g \circ f)(x) = 8x - 7$
* Easier to understand when evaluating for a specific value of x. It's the innermost value, and is plugged in to the function wrapping it. The output is plugged into the function wrapping it and so on.
 * Example: For functions $f(x) = x^2 − 4$ and $g(x) = 3x + 2$, find: $(f \circ g)(−3)$ 
   * Plug x into g: $g(-3) = 3(-3) + 2$
     * Simplify: $g(-3) = -7$
   * Plug g's output into f: $f(g(-3)) = (-7)^2 − 4$ (Notice -7 was plugged, not -3)
     * Simplify: $f(g(-3)) = 45$
   * Solution: $(f \circ g)(−3) = 45$
* Functions can also be functions of themselves.
 * Example: For function $f(x) = x^2 − 4$, find: $(f \circ f)(2)$ 
   * Plug x into f: $f(x) = (2)^2 − 4 = 0$
   * Plug output into f again: $(f \circ f)(2) = (0)^2 − 4$ (0 is the input, not 2)
   * Simplify: $(f \circ f)(2) = -4$

## Inverse Functions
* An inverse of a function is a one-to-one function that for a function $f(x)$ undoes the operation of $f$, and is denoted $f^{-1}(x)$.
* It uses composition to revert what the original function did to a value in its domain in order to get back to the original x value.
![title](./../resources/inverse-function.png)
* A function and its inverse are inverses of each other, together called **inverse functions**, so working backwards, inputting x into the inverse and then into the original function will still render the original x value.
* Inverse Functions Property:
 * $f^{-1}\left(f(x)\right) = x$, for all $x$ in the domain of $f$.
 * $f\left(f^{-1}(x)\right) = x$, for all $x$ in the domain of $f^{-1}$.
* In both functions of the property, the input is x and the output is x. When you compose a function with its inverse function that has the input of x, you get back that same value, x, making it an **identity function**.
* If an inverse function exists for a given function f, then it is unique.
* The set of ordered pair coordinates between inverse functions are **converse relations**, whereby the order of $(x,y)$ values is reversed for the inverse, becoming $(y,x)$ values. 
 * Original ordered pairs: $f(x) = {(0,5),(1,6),(2,7)}$
 * Inverse ordered pairs: $f^{-1}(x) = {(5,0),(6,1),(7,2)}$
* Thus the domain of $f$ is the range of $f^{−1}$ and the domain of $f^{−1}$ is the range of $f$.
* When graphing the points of a function & its inverse function, the graphs are mirror images of each other, equidistant from the line $y=x$, or the identity line. 

![title](./../resources/inverse-with-identity-line.png)

#### Determining If Functions Are Inverse
* As per the Inverse Functions Property, using the output of a function as the input of the inverse function, should result in x and vice versa.
* Two functions, $f(x)$ and $g(x)$, are inverses of each other if they're both the identity function, $g(f(x))=x$ and $f(g(x))=x$.
* Example: Verify that $f(x)=5x−1$ and $g(x)=\frac{x+1}{5}$ are inverse functions.
 * Plug 1st function: $f(g(x))=5(\frac{x+1}{5})−1$
   * Simplify: $f(g(x))= x+1−1$
   * $f(g(x))= x$
 * Plug 2nd function: $g(f(x))=\frac{(5x−1)+1}{5}$
   * Simplify: $g(f(x))=\frac{5x}{5}$
   * $g(f(x))= x$
 * Both = x, and are thus inverses of each other.

#### Solving for the Inverse Function
* Since a funciton's x,y coordinates are reveresed for its inverse, we can find the inverse by swapping x and y in a function's equation and then solving for y.
* Example: $f(x) = 4x + 7$
 * Substitute def: $y = 4x + 7$
 * Swap: $x = 4y + 7$
 * Solve for y: $x - 7 = 4y$
   * Isolate y: $\frac{x - 7}{4} = y$
 * Substitute inverse def: $f^{-1}(x) = \frac{x - 7}{4}$
* Example: $f(x) = \sqrt[5]{2x-3}$
 * Sub def: $y = \sqrt[5]{2x-3}$
 * Swap: $x = \sqrt[5]{2y-3}$
 * Solve for y: $x^5 = (\sqrt[5]{2y-3})^5$
   * Simplify: $x^5 = 2y-3$
   * Isolate y: $\frac{x^5 + 3}{2} = y$
 * Sub inverse def: $f{^1}(x) = \frac{x^5 + 3}{2}$
 
#### Restricting the Domain
* Some functions don't have inverses unless we restrict the domain.
* Use set builder notation to limit the domain to valid values.
* Square function does not have an inverse, but we can restrict the domain:
 * {$x^2 \mid x \geq 0$}

## Polynomial Function Operations
* A function whose range values are defined by a polynomial, i.e. a function that involves only whole number exponents of a variable in an equation. 

#### Adding/Subtracting Polynomial Functions
* When adding/subtracting functions, the range values are added/subtracted just as you would a regular polynomial. If the domain value is given, it's substituted either after combining equations or separately into each equation first. I.e. for functions $f(x)$ and $g(x)$
 * $(f+g)(x) = f(x) + g(x)$
 * $(f-g)(x) = f(x) - g(x)$
* Example: For $f(x) = 3x^2-5x+7$ and $g(x) = x^2-4x-3$ find $(f-g)(x)$.
 * Rewrite: $(f-g)(x) = (3x^2-5x+7)-(x^2-4x-3)$ (Note that no value for x was given, so we simplify.)
 * Simplify: $(f-g)(x) = 3x^2-5x+7-x^2+4x+3)$ (Note the negative was distributed.)
   * Combine Like Terms: $(f-g)(x)=2x^2-x+10$ 
* Example: For $f(x) = 3x^2-5x+7$ and $g(x) = x^2-4x-3$ find $(f-g)(-2)$.
 * Rewrite & combine (from above): $(f-g)(x) = 2x^2-x+10$
 * Substitute value: $(f-g)(-2) = 2(-2)^2-(-2)+10$
   * Solve: $(f-g)(-2) = 20$
   
#### Multiplying Polynomial Functions
* When multiplying functions, the range values are distributed just as you would a regular polynomial. If the domain value is given, it's best to substitute before to avoid lengthy distribution. I.e. for functions $f(x)$ and $g(x)$, $(f*g)(x)=f(x)*g(x)$
* Example: For $f(x) = x+2$ and $g(x) = x^2−3x−4$, find $(f*g)(2)$.
 * Rewrite: $(f*g)(2) = (x+2)(x^2−3x−4)$
 * Substitue: $(f*g)(2) = ((2)+2)((2)^2−3(2)−4)$
 * Simplify: $(f*g)(2) = (4)(-6)$
   * Solve: $(f*g)(2) = -24$
   
#### Dividing Polynomial Functions
* When dividing functions without a remainder, use the same methods as you would a regular polynomial. If the domain value is given, it's best to substitute before to avoid long division with equations. I.e. for functions $f(x)$ and $g(x)$, where $g(x)\neq0$, $\left(\frac{f}{g}\right)(x) = \frac{f(x)}{g(x)}$
* See Rational Functions section in Algebra-2 notebook for more.
* Example: For functions $f(x) = x^2−5x−14$ and $g(x)=x+2$, find $\left(\frac{f}{g}\right)(-4)$.
 * Rewrite: $\left(\frac{f}{g}\right)(x) = \frac{x^2−5x−14}{x+2}$
 * Substitute: $\left(\frac{f}{g}\right)(-4) = \frac{(-4)^2−5(-4)−14}{(-4)+2}$
 * Simplify: $\left(\frac{f}{g}\right)(-4) = \frac{22}{-2}$
   * Solve: $\left(\frac{f}{g}\right)(-4) = -11$
   
#### Finding the Remainder
* The **Remainder Theorem** states that if a polynomial function, $f(x)$, is divided by a linear factor, $x-c$, then the remainder is $f(c)$.
 * When the divisor is written as x−c, the value of the function at c, $f(c)$, is the same as the remainder from the division problem.
* Find c and then plug it into the function to find the remainder.
 * Note if the divisor has a plus operator, rather than a minus, c will be a negative in x-c form.
* Example: Divide the function $f(x) = x^3+3x+19$ by the factor $x+2$.
 * Rewrite divisor in x-c form: $x+2$ can be written as $x-(-2)$
 * Find c: $c = -2$
 * Evaluate f(c): $f(-2) = (-2)^3+3(-2)+19$
   * Solve: $f(-2) = 5$
 
#### Finding a Factor
* The **Factor Theorem** states that for any polynomial function, $f(x)$, if $x-c$ is a factor of $f(x)$, then $f(c)=0$.
* Use the same process used in the Remainder Theorem. If the remainder is 0, the divisor is a factor of the function.
* Example: Determine if $x-4$ is a factor of $f(x) = x^3-64$.
 * Rewrite divisor in x-c form: $x-4$ is already in x-c form.
 * Find c: $c=4$
 * Evaluate f(c): $f(4) = (4)^3-64$
   * Solve: $f(4) = 0$ The divisor is a factor.



## Function Real-World Applications
* There will only be one solution when evaluating the function for y, but possibly more when solving for x, e.g. quadratic.
* Use the same word problem strategies, except analyze the solutions to see which fit the situation, i.e. negative numbers may not apply to some real world scenarios.

#### Every-day Problems
* Example: The number of unread emails in Sylvia’s account is 75. This number grows by 10 unread emails a day. The function $N(t)=75+10t$ represents the relation between the number of emails, N, and the time, t, measured in days. Determine the independent & dependent variable and find $N(5)$.
 * Identify: # of emails, N, depends on # of days, t. N(5), where t=5.
 * Translate: $N(5) = 75+10(5)$
 * Evaluate: $N(5) = 125$
 
#### Number Problems
* The product of two consecutive odd integers is 323. Find the integers.
 * Identify: n and m when n * m = 323 and n + 2 = m, so n(n + 2) = 323, quadratic means 2 solutions.
 * Translate: $f(n) = n^2 + 2n$ when $f(n) = 323$ 
 * Substitue f(x) : $323 = n^2 + 2n$
 * Rewrite quadratic standard form: $n^2 + 2n - 323 = 0$
 * Factor: $(n - 17)(n + 19) = 0$
   * Solve for $n_1$: $n_1 = 17$
   * Solve for $n_2$: $n_2 = -19$
 * Find 1st solution: $m_1 = n_1 + 2$
 * Evaluate: $m_1 = 17 + 2$
   * Solve for $m_1$: $m_1 = 19$
 * Find 2nd solution: $m_2 = n_2 + 2$
 * Evaluate: $m_2 = -19 + 2$
   * Solve for $m_2$: $m_2 = -17$
 * Solutions are either 17, 19 and -17,-19
 
#### Geometry
* Example: A rectangular bedroom has an area 117 square feet. The length of the bedroom is four feet more than the width. Find the length and width of the bedroom.
 * Identify: l and w when l * w = 117 and w + 4 = l, so w(w+4) = 117, quadritic means 2 solutions, but negative numbers don't apply to some real world scenarios.
 * Translate: $w(w+4) = 117$
 * Simplify: $w^2 + 4w = 117$
 * Rewrite quatratic: $w^2 + 4w - 117 = 0$
 * Factor: $(w + 13)(w - 9) = 0$
   * Solve for $w_1, w_2$: $w_1 = -13, w_2 = 9$ (only $w_2 is positive$)
 * Solve for l: $l = (9) + 4$
   * Solve for l: $l = 13$
   
#### Physics
* Example: Polynomial functions are used in many fields to determine the height of an object at some time after it is projected into the air. The polynomial function $h(t)=−16t^2+250$ gives the height of a ball t seconds after it is dropped from a 250-foot tall building. Find the height after t=2 seconds.
 * Identify: h when 2
 * Rewrite: $h(2) = 16(2)^2+250$
 * Simplify: $h(2) = 314$
* Example: Dennis is going to throw his rubber band ball upward from the top of a campus building. When he throws the rubber band ball from 80 feet above the ground, the function $h(t)=−16t^2+64t+80$ models the height, h, of the ball above the ground as a function of time, t. Find (1)when the ball hits the ground and (2)when the ball will be 80 feet above ground.
 * 1-Identify: t when ground, ground height = 0, so $h(t) = 0$ the zero of the function.
 * 1-Translate: $0 = -16t^2 + 64t + 80$
 * 1-Rewrite quadratic: $-16t^2 + 64t + 80 = 0$
 * 1-Factor GCF: $-16(t^2 - 4t - 5) = 0$
 * 1-Factor trinomial: $-16(t-5)(t+1) = 0$
 * 1-Rewrite multiple equations: $-16\neq0, t=5, t=-1$ (16 isn't 0 and time isn't negative)
   * Solve: $h(5) = 0$
   * The ball hits the ground in 5 seconds.
 * 2-Identify: t when height 80, so $h(t) = 80$
 * 2-Translate: $80 = -16t^2 + 64t + 80$
 * 2-Rewrite quadratic: $-16t^2 + 64t + 80 - 80 = 0$
 * 2-Factor GCF: $-16t(t - 4) = 0$
 * 2-Rewrite multiple equations: $-16t=0, t-4=0$ (use zero property again on one factor of new equations and simplify the other)
 * 2-Rewrite multiple equations 1st factor: $-16\neq0,t=0$ ($-16\neq0$)
   * Solve: $h(0) = 80$
 * 2-Simplify 2nd factor: $t=4$
   * Solve: $h(4) = 80$
   * The ball will be at 80 ft the moment it's tossed and 4 seconds later as it's falling.
